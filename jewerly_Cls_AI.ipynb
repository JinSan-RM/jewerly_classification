```python
# ê°œì„ ëœ EfficientNetV2 ê¸°ë°˜ ì•…ì„¸ì„œë¦¬ ë¶„ë¥˜ AI ëª¨ë¸

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, WeightedRandomSampler
from torchvision import datasets, transforms
from torchvision.datasets import ImageFolder
import timm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from PIL import Image
import os
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ í™•ì¸ í•¨ìˆ˜
def check_available_models():
    """ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ë“¤ì„ í™•ì¸í•˜ê³  ì¶œë ¥"""
    print("ğŸ” ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ë“¤ì„ í™•ì¸ ì¤‘...")

    all_models = timm.list_models()
    efficientnet_models = [m for m in all_models if 'efficientnet' in m.lower()]

    print(f"ì´ {len(efficientnet_models)}ê°œì˜ EfficientNet ëª¨ë¸ ë°œê²¬:")

    # ì¹´í…Œê³ ë¦¬ë³„ë¡œ ì •ë¦¬
    v2_models = [m for m in efficientnet_models if 'v2' in m]
    regular_models = [m for m in efficientnet_models if 'v2' not in m]

    print("\nğŸ“‹ EfficientNetV2 ëª¨ë¸ë“¤:")
    for model in sorted(v2_models)[:10]:  # ìƒìœ„ 10ê°œë§Œ í‘œì‹œ
        print(f"  - {model}")

    print("\nğŸ“‹ ì¼ë°˜ EfficientNet ëª¨ë¸ë“¤:")
    for model in sorted(regular_models)[:10]:  # ìƒìœ„ 10ê°œë§Œ í‘œì‹œ
        print(f"  - {model}")

    return efficientnet_models

# ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ í™•ì¸
def check_available_models():
    """ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ë“¤ì„ í™•ì¸í•˜ê³  ì¶œë ¥"""
    print("ğŸ” ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ë“¤ì„ í™•ì¸ ì¤‘...")

    all_models = timm.list_models()
    efficientnet_models = [m for m in all_models if 'efficientnet' in m.lower()]

    print(f"ì´ {len(efficientnet_models)}ê°œì˜ EfficientNet ëª¨ë¸ ë°œê²¬:")

    # ì¹´í…Œê³ ë¦¬ë³„ë¡œ ì •ë¦¬
    v2_models = [m for m in efficientnet_models if 'v2' in m]
    regular_models = [m for m in efficientnet_models if 'v2' not in m]

    print("\nğŸ“‹ EfficientNetV2 ëª¨ë¸ë“¤:")
    for model in sorted(v2_models)[:10]:  # ìƒìœ„ 10ê°œë§Œ í‘œì‹œ
        print(f"  - {model}")

    print("\nğŸ“‹ ì¼ë°˜ EfficientNet ëª¨ë¸ë“¤:")
    for model in sorted(regular_models)[:10]:  # ìƒìœ„ 10ê°œë§Œ í‘œì‹œ
        print(f"  - {model}")

    return efficientnet_models

# 1. ê°œì„ ëœ ë°ì´í„° ì „ì²˜ë¦¬ ë° ì¦ê°•
class ImprovedTransforms:
    @staticmethod
    def get_train_transform():
        return transforms.Compose([
            transforms.Resize((384, 384)),  # EfficientNetV2ì— ë” ì í•©í•œ í•´ìƒë„
            transforms.RandomResizedCrop(384, scale=(0.8, 1.0)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.2),  # ì•…ì„¸ì„œë¦¬ëŠ” ìˆ˜ì§ ë°˜ì „ë„ ìœ íš¨í•  ìˆ˜ ìˆìŒ
            transforms.RandomRotation(degrees=30),
            transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.1),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1), shear=10),
            transforms.RandomPerspective(distortion_scale=0.2, p=0.3),
            transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            transforms.RandomErasing(p=0.3, scale=(0.02, 0.33), ratio=(0.3, 3.3), value=0)
        ])

    @staticmethod
    def get_val_transform():
        return transforms.Compose([
            transforms.Resize((384, 384)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

# 2. ê°œì„ ëœ EfficientNetV2 ëª¨ë¸ ì •ì˜
class ImprovedAccessoryClassifier(nn.Module):
    def __init__(self, num_classes=4, model_name=None):
        super(ImprovedAccessoryClassifier, self).__init__()

        # ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ë“¤ (ìš°ì„ ìˆœìœ„ ìˆœ)
        model_options = [
            'tf_efficientnetv2_m.in21k_ft_in1k',
            'tf_efficientnetv2_s.in21k_ft_in1k',
            'efficientnetv2_rw_m.agc_in1k',
            'efficientnetv2_rw_s.ra2_in1k',
            'efficientnet_b4.ra2_in1k',
            'efficientnet_b3.ra2_in1k'
        ]

        if model_name is None:
            # ì‚¬ìš© ê°€ëŠ¥í•œ ì²« ë²ˆì§¸ ëª¨ë¸ ì°¾ê¸°
            for model_option in model_options:
                try:
                    self.backbone = timm.create_model(model_option, pretrained=True)
                    model_name = model_option
                    print(f"âœ… ëª¨ë¸ ë¡œë“œ ì„±ê³µ: {model_name}")
                    break
                except Exception as e:
                    print(f"âŒ {model_option} ë¡œë“œ ì‹¤íŒ¨: {str(e)}")
                    continue

            if not hasattr(self, 'backbone'):
                raise RuntimeError("ì‚¬ìš© ê°€ëŠ¥í•œ EfficientNet ëª¨ë¸ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        else:
            try:
                self.backbone = timm.create_model(model_name, pretrained=True)
                print(f"âœ… ì§€ì •ëœ ëª¨ë¸ ë¡œë“œ ì„±ê³µ: {model_name}")
            except Exception as e:
                print(f"âŒ {model_name} ë¡œë“œ ì‹¤íŒ¨, ëŒ€ì•ˆ ëª¨ë¸ ì‹œë„ ì¤‘...")
                # ëŒ€ì•ˆ ëª¨ë¸ ì‹œë„
                for model_option in model_options:
                    try:
                        self.backbone = timm.create_model(model_option, pretrained=True)
                        model_name = model_option
                        print(f"âœ… ëŒ€ì•ˆ ëª¨ë¸ ë¡œë“œ ì„±ê³µ: {model_name}")
                        break
                    except:
                        continue

        # íŠ¹ì„± ì°¨ì› í™•ì¸
        backbone_features = self.backbone.num_features

        # ê¸°ì¡´ classifier ì œê±°
        self.backbone.classifier = nn.Identity()

        # ê°œì„ ëœ ë¶„ë¥˜ê¸° í—¤ë“œ
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(backbone_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

        # Xavier ì´ˆê¸°í™”
        self._initialize_weights()

        print(f"Backbone features: {backbone_features}")
        print(f"Number of classes: {num_classes}")

    def _initialize_weights(self):
        for m in self.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        features = self.backbone.forward_features(x)
        output = self.classifier(features)
        return output

# 3. ê°œì„ ëœ í•™ìŠµ í•¨ìˆ˜ë“¤
def create_weighted_sampler(dataset):
    """í´ë˜ìŠ¤ ë¶ˆê· í˜• í•´ê²°ì„ ìœ„í•œ ê°€ì¤‘ ìƒ˜í”ŒëŸ¬ ìƒì„±"""
    targets = [dataset[i][1] for i in range(len(dataset))]
    class_counts = torch.bincount(torch.tensor(targets))
    weights = 1.0 / class_counts.float()
    sample_weights = weights[targets]

    return WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(sample_weights),
        replacement=True
    )

def mixup_data(x, y, alpha=1.0):
    """Mixup ë°ì´í„° ì¦ê°•"""
    if alpha > 0:
        lam = np.random.beta(alpha, alpha)
    else:
        lam = 1

    batch_size = x.size(0)
    index = torch.randperm(batch_size).to(x.device)

    mixed_x = lam * x + (1 - lam) * x[index, :]
    y_a, y_b = y, y[index]

    return mixed_x, y_a, y_b, lam

def mixup_criterion(criterion, pred, y_a, y_b, lam):
    """Mixupì„ ìœ„í•œ ì†ì‹¤ í•¨ìˆ˜"""
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

def train_epoch_improved(model, train_loader, criterion, optimizer, device, use_mixup=True):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    progress_bar = tqdm(train_loader, desc='Training')

    for batch_idx, (images, labels) in enumerate(progress_bar):
        images, labels = images.to(device), labels.to(device)

        # Mixup ì ìš©
        if use_mixup and np.random.random() > 0.5:
            images, labels_a, labels_b, lam = mixup_data(images, labels, alpha=0.2)

            optimizer.zero_grad()
            outputs = model(images)
            loss = mixup_criterion(criterion, outputs, labels_a, labels_b, lam)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            running_loss += loss.item()

            # Mixupì—ì„œëŠ” ì •í™•ë„ ê³„ì‚°ì´ ë³µì¡í•˜ë¯€ë¡œ ë‹¨ìˆœí™”
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (lam * (predicted == labels_a).float() +
                       (1 - lam) * (predicted == labels_b).float()).sum().item()
        else:
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        # ì§„í–‰ë¥  í‘œì‹œ ì—…ë°ì´íŠ¸
        if batch_idx % 5 == 0:
            current_acc = 100. * correct / total if total > 0 else 0
            progress_bar.set_postfix({
                'Loss': f'{loss.item():.4f}',
                'Acc': f'{current_acc:.2f}%'
            })

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100. * correct / total
    return epoch_loss, epoch_acc

def validate_epoch_improved(model, val_loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc='Validation'):
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    epoch_loss = running_loss / len(val_loader)
    epoch_acc = 100. * correct / total
    return epoch_loss, epoch_acc, all_preds, all_labels

# 4. ë©”ì¸ í•™ìŠµ ì‹¤í–‰ í•¨ìˆ˜
def run_improved_training(data_path, num_epochs=50):
    # ê°œì„ ëœ ë°ì´í„° ë³€í™˜
    train_transform = ImprovedTransforms.get_train_transform()
    val_transform = ImprovedTransforms.get_val_transform()

    # ë°ì´í„°ì…‹ ë¡œë“œ
    print("ë°ì´í„°ì…‹ ë¡œë”© ì¤‘...")
    train_dataset = ImageFolder(root=os.path.join(data_path, 'train'), transform=train_transform)
    val_dataset = ImageFolder(root=os.path.join(data_path, 'val'), transform=val_transform)

    print(f"í´ë˜ìŠ¤: {train_dataset.classes}")
    print(f"í›ˆë ¨ ë°ì´í„° ìˆ˜: {len(train_dataset)}")
    print(f"ê²€ì¦ ë°ì´í„° ìˆ˜: {len(val_dataset)}")

    # í´ë˜ìŠ¤ë³„ ë°ì´í„° ê°œìˆ˜ í™•ì¸
    class_counts = {}
    for idx, (_, label) in enumerate(train_dataset):
        class_name = train_dataset.classes[label]
        class_counts[class_name] = class_counts.get(class_name, 0) + 1

    print("\ní´ë˜ìŠ¤ë³„ ë°ì´í„° ë¶„í¬:")
    for class_name, count in class_counts.items():
        print(f"  {class_name}: {count}ê°œ")

    # ê°€ì¤‘ ìƒ˜í”ŒëŸ¬ ìƒì„±
    weighted_sampler = create_weighted_sampler(train_dataset)

    # ë°ì´í„°ë¡œë” ìƒì„± (ë” í° ë°°ì¹˜ í¬ê¸°ì™€ ê°€ì¤‘ ìƒ˜í”Œë§)
    train_loader = DataLoader(
        train_dataset,
        batch_size=16,  # GPU ë©”ëª¨ë¦¬ì— ë”°ë¼ ì¡°ì •
        sampler=weighted_sampler,
        num_workers=2,
        pin_memory=True
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=16,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    # ëª¨ë¸ ìƒì„±
    print("EfficientNetV2 ëª¨ë¸ ìƒì„± ì¤‘...")
    model = ImprovedAccessoryClassifier(num_classes=len(train_dataset.classes))
    model = model.to(device)

    print(f"ì´ íŒŒë¼ë¯¸í„° ìˆ˜: {sum(p.numel() for p in model.parameters()):,}")
    print(f"í•™ìŠµ ê°€ëŠ¥í•œ íŒŒë¼ë¯¸í„° ìˆ˜: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}")

    # í´ë˜ìŠ¤ ê°€ì¤‘ì¹˜ ê³„ì‚°
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique([train_dataset[i][1] for i in range(len(train_dataset))]),
        y=[train_dataset[i][1] for i in range(len(train_dataset))]
    )
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

    # ì†ì‹¤ í•¨ìˆ˜ ë° ì˜µí‹°ë§ˆì´ì €
    criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)
    optimizer = optim.AdamW(
        model.parameters(),
        lr=0.001,  # ì´ˆê¸° í•™ìŠµë¥  ì¦ê°€
        weight_decay=0.01,
        betas=(0.9, 0.999)
    )

    # ê°œì„ ëœ ìŠ¤ì¼€ì¤„ëŸ¬
    scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
        optimizer, T_0=10, T_mult=2, eta_min=1e-6
    )

    # í•™ìŠµ ê¸°ë¡
    train_losses, train_accs = [], []
    val_losses, val_accs = [], []
    best_val_acc = 0.0
    patience_counter = 0
    early_stop_patience = 15

    print(f"\nğŸš€ EfficientNetV2 í•™ìŠµ ì‹œì‘! (ì´ {num_epochs} ì—í¬í¬)")
    print("="*70)

    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch+1}/{num_epochs}')
        print('-' * 40)

        # í•™ìŠµ
        train_loss, train_acc = train_epoch_improved(
            model, train_loader, criterion, optimizer, device, use_mixup=True
        )

        # ê²€ì¦
        val_loss, val_acc, val_preds, val_labels = validate_epoch_improved(
            model, val_loader, criterion, device
        )

        # ìŠ¤ì¼€ì¤„ëŸ¬ ì—…ë°ì´íŠ¸
        scheduler.step()
        current_lr = optimizer.param_groups[0]['lr']

        # ê²°ê³¼ ì €ì¥
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print(f'Learning Rate: {current_lr:.2e}')

        # ìµœê³  ì„±ëŠ¥ ëª¨ë¸ ì €ì¥
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'best_efficientnetv2_accessory_model.pth')
            print(f'ğŸ‰ ìƒˆë¡œìš´ ìµœê³  ì„±ëŠ¥! ëª¨ë¸ ì €ì¥ë¨ (Val Acc: {best_val_acc:.2f}%)')
            patience_counter = 0

            # ìµœê³  ì„±ëŠ¥ ë‹¬ì„± ì‹œ ë¶„ë¥˜ ë³´ê³ ì„œ ì¶œë ¥
            if epoch > 5:  # ì´ˆê¸° ëª‡ ì—í¬í¬ëŠ” ì œì™¸
                print("\nğŸ“Š í˜„ì¬ ìµœê³  ì„±ëŠ¥ ëª¨ë¸ì˜ í´ë˜ìŠ¤ë³„ ì„±ëŠ¥:")
                print(classification_report(
                    val_labels, val_preds,
                    target_names=train_dataset.classes,
                    digits=3
                ))
        else:
            patience_counter += 1

        # Early stopping
        if patience_counter >= early_stop_patience:
            print(f"\nâ° Early stopping! {early_stop_patience} ì—í¬í¬ ë™ì•ˆ ê°œì„ ì´ ì—†ì—ˆìŠµë‹ˆë‹¤.")
            break

        # 10 ì—í¬í¬ë§ˆë‹¤ ì¤‘ê°„ ê²°ê³¼ ì‹œê°í™”
        if (epoch + 1) % 10 == 0:
            plot_training_progress(train_losses, train_accs, val_losses, val_accs, epoch + 1)

    print(f'\nğŸ¯ í•™ìŠµ ì™„ë£Œ! ìµœê³  ê²€ì¦ ì •í™•ë„: {best_val_acc:.2f}%')

    return model, train_losses, train_accs, val_losses, val_accs, best_val_acc

def plot_training_progress(train_losses, train_accs, val_losses, val_accs, current_epoch):
    """í•™ìŠµ ì§„í–‰ ìƒí™© ì‹œê°í™”"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    epochs = range(1, len(train_losses) + 1)

    # Loss ê·¸ë˜í”„
    axes[0].plot(epochs, train_losses, 'b-', label='Train Loss', linewidth=2)
    axes[0].plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2)
    axes[0].set_title(f'Training Progress - Loss (Epoch {current_epoch})')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Accuracy ê·¸ë˜í”„
    axes[1].plot(epochs, train_accs, 'b-', label='Train Accuracy', linewidth=2)
    axes[1].plot(epochs, val_accs, 'r-', label='Validation Accuracy', linewidth=2)
    axes[1].set_title(f'Training Progress - Accuracy (Epoch {current_epoch})')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy (%)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    # ìµœê³  ì„±ëŠ¥ í‘œì‹œ
    best_epoch = np.argmax(val_accs) + 1
    best_acc = max(val_accs)
    axes[1].axvline(x=best_epoch, color='green', linestyle='--', alpha=0.7)
    axes[1].text(best_epoch, best_acc-5, f'Best: {best_acc:.1f}%\nEpoch {best_epoch}',
                ha='center', va='top', fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", facecolor='yellow', alpha=0.7))

    plt.tight_layout()
    plt.show()

# 5. ì˜ˆì¸¡ í•¨ìˆ˜
def predict_accessory_improved(image_path, model, transform, device, class_names):
    """ê°œì„ ëœ ì˜ˆì¸¡ í•¨ìˆ˜ (TTA ì ìš©)"""
    model.eval()

    # ì´ë¯¸ì§€ ë¡œë“œ
    image = Image.open(image_path).convert('RGB')

    # TTA (Test Time Augmentation)
    predictions = []

    # ì›ë³¸ ì´ë¯¸ì§€
    image_tensor = transform(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(image_tensor)
        predictions.append(F.softmax(output, dim=1).cpu())

    # ìˆ˜í‰ ë°˜ì „
    image_flipped = image.transpose(Image.FLIP_LEFT_RIGHT)
    image_tensor = transform(image_flipped).unsqueeze(0).to(device)
    with torch.no_grad():
        output = model(image_tensor)
        predictions.append(F.softmax(output, dim=1).cpu())

    # í‰ê·  ì˜ˆì¸¡
    avg_prediction = torch.mean(torch.stack(predictions), dim=0)
    predicted_class = torch.argmax(avg_prediction, dim=1).item()
    confidence = avg_prediction[0][predicted_class].item()

    return class_names[predicted_class], confidence, avg_prediction[0].numpy()

# ì‚¬ìš© ì˜ˆì‹œ
if __name__ == "__main__":
    # GPU ì„¤ì •
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f'ì‚¬ìš© ë””ë°”ì´ìŠ¤: {device}')

    # ì‚¬ìš© ê°€ëŠ¥í•œ ëª¨ë¸ í™•ì¸ (ì„ íƒì‚¬í•­)
    # available_models = check_available_models()

    # ë°ì´í„° ê²½ë¡œ ì„¤ì • (ì‹¤ì œ ê²½ë¡œë¡œ ë³€ê²½ í•„ìš”)
    data_path = '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/accesary_datasets'

    # ê°œì„ ëœ í•™ìŠµ ì‹¤í–‰
    if os.path.exists(data_path):
        model, train_losses, train_accs, val_losses, val_accs, best_acc = run_improved_training(
            data_path, num_epochs=50
        )

        print(f"\nğŸ¯ ìµœì¢… ê²°ê³¼:")
        print(f"ìµœê³  ê²€ì¦ ì •í™•ë„: {best_acc:.2f}%")
        print(f"ëª¨ë¸ì´ 'best_efficientnetv2_accessory_model.pth'ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

        # ìµœì¢… ì‹œê°í™”
        plot_training_progress(train_losses, train_accs, val_losses, val_accs, len(train_losses))

    else:
        print(f"ë°ì´í„° ê²½ë¡œ '{data_path}'ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
        print("ë°ì´í„° ê²½ë¡œë¥¼ í™•ì¸í•˜ê±°ë‚˜ ë°ì´í„°ë¥¼ ì—…ë¡œë“œí•´ì£¼ì„¸ìš”.")
```

```python
# ì§„ì§œ ë©”ëª¨ë¦¬ ìµœì í™”ëœ ê³ ì„±ëŠ¥ ì£¼ì–¼ë¦¬ ë¶„ë¥˜ ëª¨ë¸
# GPU ë©”ëª¨ë¦¬ ë¶€ì¡± ë¬¸ì œ ì™„ì „ í•´ê²° + ë†’ì€ ì„±ëŠ¥ ìœ ì§€

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, WeightedRandomSampler
from torch.cuda.amp import GradScaler, autocast
from torchvision import transforms
from torchvision.datasets import ImageFolder
import timm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from PIL import Image
import os
from tqdm import tqdm
import warnings
import gc
warnings.filterwarnings('ignore')

# 1. GPU ë©”ëª¨ë¦¬ ìµœì í™” ì´ˆê¸° ì„¤ì •
def setup_memory_optimization():
    """GPU ë©”ëª¨ë¦¬ ìµœì í™” ì„¤ì •"""
    if torch.cuda.is_available():
        # ë©”ëª¨ë¦¬ ì¡°ê°í™” ë°©ì§€
        os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

        # CUDA ì„¤ì • ìµœì í™”
        torch.backends.cudnn.benchmark = True
        torch.backends.cudnn.deterministic = False
        torch.backends.cudnn.enabled = True

        # ë©”ëª¨ë¦¬ ìºì‹œ ì •ë¦¬
        torch.cuda.empty_cache()
        gc.collect()

        # GPU ì •ë³´ ì¶œë ¥
        device_props = torch.cuda.get_device_properties(0)
        total_memory = device_props.total_memory / 1024**3
        allocated_memory = torch.cuda.memory_allocated() / 1024**3

        print(f"ğŸš€ GPU ë©”ëª¨ë¦¬ ìµœì í™” ì™„ë£Œ")
        print(f"ğŸ“Š ì´ GPU ë©”ëª¨ë¦¬: {total_memory:.1f} GB")
        print(f"ğŸ’¾ í˜„ì¬ ì‚¬ìš©ëŸ‰: {allocated_memory:.1f} GB")
        print(f"ğŸ†“ ì‚¬ìš© ê°€ëŠ¥: {total_memory - allocated_memory:.1f} GB")

        return total_memory
    return 0

# 2. ì ì‘ì  ì„¤ì • í´ë˜ìŠ¤
class AdaptiveConfig:
    def __init__(self, gpu_memory_gb):
        self.gpu_memory = gpu_memory_gb

        if gpu_memory_gb < 6:
            # ë‚®ì€ ë©”ëª¨ë¦¬ ì„¤ì •
            self.image_size = 224
            self.batch_size = 4
            self.accumulation_steps = 8
            self.num_workers = 1
            self.model_size = 'small'
        elif gpu_memory_gb < 10:
            # ì¤‘ê°„ ë©”ëª¨ë¦¬ ì„¤ì •
            self.image_size = 256
            self.batch_size = 6
            self.accumulation_steps = 6
            self.num_workers = 2
            self.model_size = 'medium'
        elif gpu_memory_gb < 14:
            # ë†’ì€ ë©”ëª¨ë¦¬ ì„¤ì •
            self.image_size = 288
            self.batch_size = 8
            self.accumulation_steps = 4
            self.num_workers = 2
            self.model_size = 'large'
        else:
            # ìµœëŒ€ ë©”ëª¨ë¦¬ ì„¤ì •
            self.image_size = 320
            self.batch_size = 12
            self.accumulation_steps = 2
            self.num_workers = 3
            self.model_size = 'xlarge'

        # ì‹¤ì œ ë°°ì¹˜ í¬ê¸°ëŠ” batch_size * accumulation_steps
        self.effective_batch_size = self.batch_size * self.accumulation_steps

        print(f"ğŸ”§ ì ì‘ì  ì„¤ì • (GPU {gpu_memory_gb:.1f}GB ê¸°ì¤€):")
        print(f"   ì´ë¯¸ì§€ í¬ê¸°: {self.image_size}x{self.image_size}")
        print(f"   ë¬¼ë¦¬ì  ë°°ì¹˜: {self.batch_size}")
        print(f"   ê°€ìƒ ë°°ì¹˜: {self.effective_batch_size}")
        print(f"   ëˆ„ì  ë‹¨ê³„: {self.accumulation_steps}")
        print(f"   ì›Œì»¤ ìˆ˜: {self.num_workers}")
        print(f"   ëª¨ë¸ í¬ê¸°: {self.model_size}")

# 3. ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ë°ì´í„° ë³€í™˜
class MemoryOptimizedTransforms:
    @staticmethod
    def get_train_transform(image_size):
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.RandomResizedCrop(image_size, scale=(0.8, 1.0)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.3),
            transforms.RandomRotation(degrees=20),  # ë©”ëª¨ë¦¬ ì ˆì•½ì„ ìœ„í•´ ê°ë„ ì¶•ì†Œ
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            transforms.RandomErasing(p=0.25, scale=(0.02, 0.2), ratio=(0.3, 3.3)),
        ])

    @staticmethod
    def get_val_transform(image_size):
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

# 4. ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ëª¨ë¸ í´ë˜ìŠ¤
class MemoryEfficientClassifier(nn.Module):
    def __init__(self, num_classes=4, model_size='medium'):
        super(MemoryEfficientClassifier, self).__init__()

        # ëª¨ë¸ í¬ê¸°ë³„ ë°±ë³¸ ì„ íƒ
        model_configs = {
            'small': [
                'mobilenetv3_large_100.ra_in1k',
                'efficientnet_b2.ra_in1k',
                'resnet50.a1_in1k'
            ],
            'medium': [
                'efficientnet_b3.ra2_in1k',
                'efficientnet_b4.ra2_in1k',
                'efficientnetv2_rw_s.ra2_in1k'
            ],
            'large': [
                'efficientnetv2_rw_m.agc_in1k',
                'efficientnet_b5.ra_in1k',
                'convnext_small.fb_in22k_ft_in1k'
            ],
            'xlarge': [
                'efficientnetv2_m.in21k_ft_in1k',
                'convnext_base.fb_in22k_ft_in1k',
                'efficientnet_b6.ra_in1k',
            ]
        }

        # ë°±ë³¸ ë¡œë“œ
        self.backbone = None
        backbone_features = None

        for model_name in model_configs[model_size]:
            try:
                self.backbone = timm.create_model(model_name, pretrained=True)
                backbone_features = self.backbone.num_features
                self.backbone.classifier = nn.Identity()

                # Gradient Checkpointing í™œì„±í™” (ë©”ëª¨ë¦¬ 50% ì ˆì•½)
                if hasattr(self.backbone, 'set_grad_checkpointing'):
                    self.backbone.set_grad_checkpointing(True)

                print(f"âœ… ë©”ëª¨ë¦¬ íš¨ìœ¨ ëª¨ë¸ ë¡œë“œ: {model_name}")
                break
            except Exception as e:
                print(f"âŒ {model_name} ë¡œë“œ ì‹¤íŒ¨, ë‹¤ìŒ ëª¨ë¸ ì‹œë„ ì¤‘...")
                continue

        if self.backbone is None:
            raise RuntimeError("ì‚¬ìš© ê°€ëŠ¥í•œ ëª¨ë¸ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        # ëª¨ë¸ í¬ê¸°ë³„ ë¶„ë¥˜ê¸° í—¤ë“œ ì„¤ì •
        if model_size == 'small':
            hidden_size = 256
        elif model_size == 'medium':
            hidden_size = 512
        elif model_size == 'large':
            hidden_size = 768
        else:  # xlarge
            hidden_size = 1024

        # ê²½ëŸ‰í™”ëœ ë¶„ë¥˜ê¸° í—¤ë“œ
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(backbone_features, hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.BatchNorm1d(hidden_size // 2),
            nn.ReLU(inplace=True),
            nn.Dropout(0.1),
            nn.Linear(hidden_size // 2, num_classes)
        )

        # ê°€ì¤‘ì¹˜ ì´ˆê¸°í™”
        self._initialize_weights()

        # íŒŒë¼ë¯¸í„° ìˆ˜ ì¶œë ¥
        total_params = sum(p.numel() for p in self.parameters())
        trainable_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        print(f"ğŸ“Š ëª¨ë¸ íŒŒë¼ë¯¸í„°: {total_params:,} (í•™ìŠµê°€ëŠ¥: {trainable_params:,})")

    def _initialize_weights(self):
        for m in self.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)

    def forward(self, x):
        # Gradient checkpointingìœ¼ë¡œ ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ forward
        features = self.backbone.forward_features(x)
        output = self.classifier(features)
        return output

# 5. ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ íŠ¸ë ˆì´ë„ˆ í´ë˜ìŠ¤
class MemoryEfficientTrainer:
    def __init__(self, model, criterion, optimizer, scheduler, device, config):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device
        self.config = config

        # Mixed Precision Scaler
        self.scaler = GradScaler()

        print(f"ğŸ”¥ ë©”ëª¨ë¦¬ íš¨ìœ¨ íŠ¸ë ˆì´ë„ˆ ì´ˆê¸°í™” ì™„ë£Œ")
        print(f"   ì‹¤ì œ ë°°ì¹˜ í¬ê¸°: {config.effective_batch_size}")
        print(f"   ëˆ„ì  ë‹¨ê³„: {config.accumulation_steps}")

    def train_epoch(self, train_loader, epoch, use_mixup=True):
        """ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ í•™ìŠµ ì—í¬í¬"""
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        # ì˜µí‹°ë§ˆì´ì € ì´ˆê¸°í™”
        self.optimizer.zero_grad()

        progress_bar = tqdm(train_loader, desc=f'ğŸ”¥ Epoch {epoch}')

        for batch_idx, (images, labels) in enumerate(progress_bar):
            # ë°ì´í„°ë¥¼ GPUë¡œ ì´ë™ (non_blocking=Trueë¡œ ì„±ëŠ¥ í–¥ìƒ)
            images = images.to(self.device, non_blocking=True)
            labels = labels.to(self.device, non_blocking=True)

            # Mixed Precision Training
            with autocast():
                # Mixup ì ìš© (50% í™•ë¥ )
                if use_mixup and np.random.random() > 0.5:
                    images, labels_a, labels_b, lam = self._mixup_data(images, labels, alpha=0.2)
                    outputs = self.model(images)
                    loss = lam * self.criterion(outputs, labels_a) + (1 - lam) * self.criterion(outputs, labels_b)
                else:
                    outputs = self.model(images)
                    loss = self.criterion(outputs, labels)

                # Gradient Accumulationì„ ìœ„í•œ loss ìŠ¤ì¼€ì¼ë§
                loss = loss / self.config.accumulation_steps

            # Scaled backward pass
            self.scaler.scale(loss).backward()

            # í†µê³„ ì—…ë°ì´íŠ¸
            running_loss += loss.item() * self.config.accumulation_steps
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # Gradient Accumulation ë° ì—…ë°ì´íŠ¸
            if (batch_idx + 1) % self.config.accumulation_steps == 0:
                # Gradient Clipping
                self.scaler.unscale_(self.optimizer)
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)

                # ì˜µí‹°ë§ˆì´ì € ìŠ¤í…
                self.scaler.step(self.optimizer)
                self.scaler.update()
                self.optimizer.zero_grad()

                # ì •ê¸°ì  ë©”ëª¨ë¦¬ ì •ë¦¬ (ë§¤ 20 ëˆ„ì  ìŠ¤í…ë§ˆë‹¤)
                if (batch_idx + 1) % (self.config.accumulation_steps * 20) == 0:
                    torch.cuda.empty_cache()

            # ì§„í–‰ë¥  ì—…ë°ì´íŠ¸
            if batch_idx % 10 == 0:
                current_acc = 100. * correct / total if total > 0 else 0
                allocated_memory = torch.cuda.memory_allocated() / 1024**3
                progress_bar.set_postfix({
                    'Loss': f'{loss.item() * self.config.accumulation_steps:.4f}',
                    'Acc': f'{current_acc:.2f}%',
                    'GPU': f'{allocated_memory:.1f}GB'
                })

            # ë©”ëª¨ë¦¬ ëˆ„ìˆ˜ ë°©ì§€
            del images, labels, outputs, loss

        # ë§ˆì§€ë§‰ ë°°ì¹˜ ì²˜ë¦¬
        if (batch_idx + 1) % self.config.accumulation_steps != 0:
            self.scaler.unscale_(self.optimizer)
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            self.scaler.step(self.optimizer)
            self.scaler.update()
            self.optimizer.zero_grad()

        # ìŠ¤ì¼€ì¤„ëŸ¬ ì—…ë°ì´íŠ¸
        self.scheduler.step()

        # ìµœì¢… ë©”ëª¨ë¦¬ ì •ë¦¬
        torch.cuda.empty_cache()
        gc.collect()

        epoch_loss = running_loss / len(train_loader)
        epoch_acc = 100. * correct / total

        return epoch_loss, epoch_acc

    def validate_epoch(self, val_loader):
        """ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ê²€ì¦ ì—í¬í¬"""
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, labels in tqdm(val_loader, desc='ğŸ” Validation'):
                images = images.to(self.device, non_blocking=True)
                labels = labels.to(self.device, non_blocking=True)

                # Mixed Precision Inference
                with autocast():
                    outputs = self.model(images)
                    loss = self.criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                # ë©”ëª¨ë¦¬ ì •ë¦¬
                del images, labels, outputs, loss

        # ê²€ì¦ ì™„ë£Œ í›„ ë©”ëª¨ë¦¬ ì •ë¦¬
        torch.cuda.empty_cache()

        epoch_loss = running_loss / len(val_loader)
        epoch_acc = 100. * correct / total

        return epoch_loss, epoch_acc, all_preds, all_labels

    def _mixup_data(self, x, y, alpha=1.0):
        """Mixup ë°ì´í„° ì¦ê°•"""
        if alpha > 0:
            lam = np.random.beta(alpha, alpha)
        else:
            lam = 1

        batch_size = x.size(0)
        index = torch.randperm(batch_size).to(x.device)

        mixed_x = lam * x + (1 - lam) * x[index, :]
        y_a, y_b = y, y[index]

        return mixed_x, y_a, y_b, lam

# 6. ë°ì´í„°ë¡œë” ìƒì„± í•¨ìˆ˜
def create_memory_efficient_dataloaders(data_path, config):
    """ë©”ëª¨ë¦¬ íš¨ìœ¨ì ì¸ ë°ì´í„°ë¡œë” ìƒì„±"""

    # ë°ì´í„° ë³€í™˜
    train_transform = MemoryOptimizedTransforms.get_train_transform(config.image_size)
    val_transform = MemoryOptimizedTransforms.get_val_transform(config.image_size)

    # ë°ì´í„°ì…‹ ë¡œë“œ
    train_dataset = ImageFolder(root=os.path.join(data_path, 'train'), transform=train_transform)
    val_dataset = ImageFolder(root=os.path.join(data_path, 'val'), transform=val_transform)

    print(f"ğŸ“ ë°ì´í„°ì…‹ ë¡œë”© ì™„ë£Œ:")
    print(f"   í›ˆë ¨ ë°ì´í„°: {len(train_dataset)}ê°œ")
    print(f"   ê²€ì¦ ë°ì´í„°: {len(val_dataset)}ê°œ")
    print(f"   í´ë˜ìŠ¤: {train_dataset.classes}")

    # í´ë˜ìŠ¤ ê· í˜• í™•ì¸
    class_counts = {}
    for i in range(len(train_dataset)):
        _, label = train_dataset[i]
        class_name = train_dataset.classes[label]
        class_counts[class_name] = class_counts.get(class_name, 0) + 1

    print(f"\nğŸ“Š í´ë˜ìŠ¤ë³„ ë°ì´í„° ë¶„í¬:")
    for class_name, count in class_counts.items():
        print(f"   {class_name}: {count}ê°œ")

    # ê°€ì¤‘ ìƒ˜í”ŒëŸ¬ ìƒì„±
    def create_weighted_sampler(dataset):
        targets = [dataset[i][1] for i in range(len(dataset))]
        class_counts = torch.bincount(torch.tensor(targets))
        weights = 1.0 / class_counts.float()
        sample_weights = weights[targets]
        return WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

    weighted_sampler = create_weighted_sampler(train_dataset)

    # ë°ì´í„°ë¡œë” ìƒì„±
    train_loader = DataLoader(
        train_dataset,
        batch_size=config.batch_size,
        sampler=weighted_sampler,
        num_workers=config.num_workers,
        pin_memory=True,
        persistent_workers=True if config.num_workers > 0 else False,
        prefetch_factor=2 if config.num_workers > 0 else None
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=config.batch_size * 2,  # ê²€ì¦ì‹œì—ëŠ” ë” í° ë°°ì¹˜ ì‚¬ìš© ê°€ëŠ¥
        shuffle=False,
        num_workers=config.num_workers,
        pin_memory=True,
        persistent_workers=True if config.num_workers > 0 else False,
        prefetch_factor=2 if config.num_workers > 0 else None
    )

    return train_loader, val_loader, train_dataset.classes

# 7. ë©”ì¸ í•™ìŠµ í•¨ìˆ˜
def run_memory_optimized_training(data_path, num_epochs=50):
    """ë©”ëª¨ë¦¬ ìµœì í™”ëœ í•™ìŠµ ì‹¤í–‰"""

    print("ğŸš€ ì§„ì§œ ë©”ëª¨ë¦¬ ìµœì í™” ê³ ì„±ëŠ¥ í•™ìŠµ ì‹œì‘!")
    print("=" * 80)

    # 1. ë©”ëª¨ë¦¬ ìµœì í™” ì„¤ì •
    gpu_memory = setup_memory_optimization()
    config = AdaptiveConfig(gpu_memory)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # 2. ë°ì´í„°ë¡œë” ìƒì„±
    train_loader, val_loader, class_names = create_memory_efficient_dataloaders(data_path, config)

    # 3. ëª¨ë¸ ìƒì„±
    print(f"\nğŸ¤– ë©”ëª¨ë¦¬ íš¨ìœ¨ ëª¨ë¸ ìƒì„± ì¤‘...")
    model = MemoryEfficientClassifier(
        num_classes=len(class_names),
        model_size=config.model_size
    ).to(device)

    # 4. í´ë˜ìŠ¤ ê°€ì¤‘ì¹˜ ê³„ì‚°
    train_dataset = train_loader.dataset
    class_weights = compute_class_weight(
        'balanced',
        classes=np.unique([train_dataset[i][1] for i in range(len(train_dataset))]),
        y=[train_dataset[i][1] for i in range(len(train_dataset))]
    )
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

    # 5. ì†ì‹¤í•¨ìˆ˜, ì˜µí‹°ë§ˆì´ì €, ìŠ¤ì¼€ì¤„ëŸ¬
    criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)

    # ì ì‘ì  í•™ìŠµë¥  ì„¤ì •
    base_lr = 0.001
    if config.model_size == 'small':
        learning_rate = base_lr * 0.8
    elif config.model_size == 'xlarge':
        learning_rate = base_lr * 1.2
    else:
        learning_rate = base_lr

    optimizer = optim.AdamW(
        model.parameters(),
        lr=learning_rate,
        weight_decay=0.01,
        betas=(0.9, 0.999)
    )

    scheduler = optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=learning_rate * 3,
        epochs=num_epochs,
        steps_per_epoch=len(train_loader) // config.accumulation_steps,
        pct_start=0.1,
        anneal_strategy='cos'
    )

    # 6. íŠ¸ë ˆì´ë„ˆ ì´ˆê¸°í™”
    trainer = MemoryEfficientTrainer(model, criterion, optimizer, scheduler, device, config)

    # 7. í•™ìŠµ ë£¨í”„
    train_losses, train_accs = [], []
    val_losses, val_accs = [], []
    best_val_acc = 0.0
    patience_counter = 0
    early_stop_patience = 15

    print(f"\nğŸ¯ í•™ìŠµ ì‹œì‘ (ì—í¬í¬: {num_epochs})")
    print(f"ğŸ“Š ì„¤ì • ìš”ì•½:")
    print(f"   ì´ë¯¸ì§€ í¬ê¸°: {config.image_size}x{config.image_size}")
    print(f"   ì‹¤ì œ ë°°ì¹˜ í¬ê¸°: {config.effective_batch_size}")
    print(f"   í•™ìŠµë¥ : {learning_rate:.2e}")
    print("=" * 60)

    for epoch in range(num_epochs):
        print(f'\nğŸ“… Epoch {epoch+1}/{num_epochs}')
        print('-' * 50)

        # GPU ë©”ëª¨ë¦¬ ìƒíƒœ í™•ì¸ (5 ì—í¬í¬ë§ˆë‹¤)
        if epoch % 5 == 0:
            allocated = torch.cuda.memory_allocated() / 1024**3
            reserved = torch.cuda.memory_reserved() / 1024**3
            print(f"ğŸ’¾ GPU ë©”ëª¨ë¦¬ ìƒíƒœ - í• ë‹¹: {allocated:.1f}GB, ì˜ˆì•½: {reserved:.1f}GB")

        # í•™ìŠµ
        train_loss, train_acc = trainer.train_epoch(train_loader, epoch, use_mixup=True)

        # ê²€ì¦
        val_loss, val_acc, val_preds, val_labels = trainer.validate_epoch(val_loader)

        # ê²°ê³¼ ì €ì¥
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        current_lr = optimizer.param_groups[0]['lr']

        print(f'ğŸ“Š Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'ğŸ“Š Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print(f'ğŸ¯ Learning Rate: {current_lr:.2e}')

        # ìµœê³  ì„±ëŠ¥ ëª¨ë¸ ì €ì¥
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/memory_optimized_best_model.pth')
            print(f'ğŸ‰ ìƒˆë¡œìš´ ìµœê³  ì„±ëŠ¥! ëª¨ë¸ ì €ì¥ë¨ (Val Acc: {best_val_acc:.2f}%)')
            patience_counter = 0

            # ìƒì„¸ ì„±ëŠ¥ ì¶œë ¥ (5 ì—í¬í¬ í›„ë¶€í„°)
            if epoch >= 5:
                print("\nğŸ“ˆ í´ë˜ìŠ¤ë³„ ì„±ëŠ¥:")
                report = classification_report(
                    val_labels, val_preds,
                    target_names=class_names,
                    digits=3, output_dict=True
                )
                for class_name in class_names:
                    metrics = report[class_name]
                    print(f"   {class_name}: P={metrics['precision']:.3f}, R={metrics['recall']:.3f}, F1={metrics['f1-score']:.3f}")
        else:
            patience_counter += 1

        # Early stopping
        if patience_counter >= early_stop_patience:
            print(f"\nâ° Early stopping! {early_stop_patience} ì—í¬í¬ ë™ì•ˆ ê°œì„  ì—†ìŒ")
            break

        # 10 ì—í¬í¬ë§ˆë‹¤ ì¤‘ê°„ ì‹œê°í™”
        if (epoch + 1) % 10 == 0:
            plot_training_progress(train_losses, train_accs, val_losses, val_accs, epoch + 1)

    print(f'\nğŸ¯ ë©”ëª¨ë¦¬ ìµœì í™” í•™ìŠµ ì™„ë£Œ!')
    print(f'ğŸ† ìµœê³  ê²€ì¦ ì •í™•ë„: {best_val_acc:.2f}%')

    # ìµœì¢… ë©”ëª¨ë¦¬ ì •ë¦¬
    torch.cuda.empty_cache()
    gc.collect()

    return model, train_losses, train_accs, val_losses, val_accs, best_val_acc, class_names

# 8. ì‹œê°í™” í•¨ìˆ˜
def plot_training_progress(train_losses, train_accs, val_losses, val_accs, current_epoch):
    """í•™ìŠµ ì§„í–‰ ìƒí™© ì‹œê°í™”"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    epochs = range(1, len(train_losses) + 1)

    # Loss ê·¸ë˜í”„
    axes[0].plot(epochs, train_losses, 'b-', label='Train Loss', linewidth=2.5)
    axes[0].plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2.5)
    axes[0].fill_between(epochs, train_losses, alpha=0.3, color='blue')
    axes[0].fill_between(epochs, val_losses, alpha=0.3, color='red')
    axes[0].set_title(f'ğŸ“‰ Training Progress - Loss (Epoch {current_epoch})', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # Accuracy ê·¸ë˜í”„
    axes[1].plot(epochs, train_accs, 'b-', label='Train Accuracy', linewidth=2.5)
    axes[1].plot(epochs, val_accs, 'r-', label='Validation Accuracy', linewidth=2.5)
    axes[1].fill_between(epochs, train_accs, alpha=0.3, color='blue')
    axes[1].fill_between(epochs, val_accs, alpha=0.3, color='red')
    axes[1].set_title(f'ğŸ“ˆ Training Progress - Accuracy (Epoch {current_epoch})', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy (%)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    # ìµœê³  ì„±ëŠ¥ í‘œì‹œ
    if val_accs:
        best_epoch = np.argmax(val_accs) + 1
        best_acc = max(val_accs)
        axes[1].axvline(x=best_epoch, color='green', linestyle='--', alpha=0.8, linewidth=2)
        axes[1].text(best_epoch, best_acc-3, f'ğŸ† Best: {best_acc:.1f}%\nEpoch {best_epoch}',
                    ha='center', va='top', fontsize=11, fontweight='bold',
                    bbox=dict(boxstyle="round,pad=0.4", facecolor='yellow', alpha=0.8, edgecolor='orange'))

    plt.tight_layout()
    plt.show()

    # í˜„ì¬ ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ ì¶œë ¥
    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**3
        reserved = torch.cuda.memory_reserved() / 1024**3
        print(f"ğŸ’¾ í˜„ì¬ GPU ë©”ëª¨ë¦¬ - í• ë‹¹: {allocated:.1f}GB, ì˜ˆì•½: {reserved:.1f}GB")

# 9. TTA ì ìš© ìµœì¢… í‰ê°€ í•¨ìˆ˜
def evaluate_with_tta(model, val_loader, class_names, device, num_tta=5):
    """TTA ì ìš©í•œ ìµœì¢… ëª¨ë¸ í‰ê°€"""

    print("\nğŸ” TTA ì ìš© ìµœì¢… í‰ê°€ ì‹œì‘")
    print("=" * 50)

    model.eval()
    all_preds = []
    all_labels = []
    all_probs = []

    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc='ğŸ¯ TTA Evaluation'):
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)

            # TTA ì˜ˆì¸¡
            predictions = []

            # ì›ë³¸
            with autocast():
                outputs = model(images)
                predictions.append(F.softmax(outputs, dim=1))

            # TTA ë³€í™˜ë“¤ (ë©”ëª¨ë¦¬ íš¨ìœ¨ì ìœ¼ë¡œ)
            tta_transforms = [
                lambda x: torch.flip(x, dims=[3]),  # ìˆ˜í‰ ë°˜ì „
                lambda x: torch.flip(x, dims=[2]),  # ìˆ˜ì§ ë°˜ì „
                lambda x: torch.rot90(x, k=2, dims=[2, 3]),  # 180ë„ íšŒì „
            ]

            for i, transform in enumerate(tta_transforms[:num_tta-1]):
                try:
                    transformed = transform(images.clone())
                    with autocast():
                        outputs = model(transformed)
                        predictions.append(F.softmax(outputs, dim=1))
                    del transformed, outputs
                except:
                    continue

            # í‰ê·  ì˜ˆì¸¡
            if predictions:
                avg_prediction = torch.mean(torch.stack(predictions), dim=0)
                _, predicted = torch.max(avg_prediction, 1)

                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_probs.extend(avg_prediction.cpu().numpy())

            # ë©”ëª¨ë¦¬ ì •ë¦¬
            del images, labels, predictions
            if torch.cuda.is_available():
                torch.cuda.empty_cache()

    # ì„±ëŠ¥ ê³„ì‚°
    final_accuracy = 100 * np.mean(np.array(all_preds) == np.array(all_labels))

    print(f"\nğŸ¯ ìµœì¢… TTA ì ìš© ì •í™•ë„: {final_accuracy:.2f}%")

    # ìƒì„¸ ë¶„ì„
    from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

    precision, recall, f1, support = precision_recall_fscore_support(all_labels, all_preds, average=None)
    cm = confusion_matrix(all_labels, all_preds)

    print(f"\nğŸ“ˆ í´ë˜ìŠ¤ë³„ ìµœì¢… ì„±ëŠ¥:")
    print("-" * 65)
    print(f"{'í´ë˜ìŠ¤':<12} | {'ì •ë°€ë„':<8} | {'ì¬í˜„ìœ¨':<8} | {'F1ì ìˆ˜':<8} | {'ìƒ˜í”Œìˆ˜':<6}")
    print("-" * 65)

    for i, class_name in enumerate(class_names):
        print(f"{class_name:<12} | {precision[i]:<8.3f} | {recall[i]:<8.3f} | {f1[i]:<8.3f} | {support[i]:<6}")

    print("-" * 65)
    print(f"{'í‰ê· ':<12} | {np.mean(precision):<8.3f} | {np.mean(recall):<8.3f} | {np.mean(f1):<8.3f} | {np.sum(support):<6}")

    # í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names,
                cbar_kws={'label': 'Count'})
    plt.title(f'ğŸ¯ Confusion Matrix\nFinal Accuracy: {final_accuracy:.2f}%', fontsize=16, fontweight='bold')
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.tight_layout()
    plt.show()

    # ì‹ ë¢°ë„ ë¶„ì„
    max_probs = np.max(all_probs, axis=1)
    correct_mask = np.array(all_preds) == np.array(all_labels)

    print(f"\nğŸ” ì˜ˆì¸¡ ì‹ ë¢°ë„ ë¶„ì„:")
    print(f"   ì „ì²´ í‰ê·  ì‹ ë¢°ë„: {np.mean(max_probs):.3f}")
    print(f"   ì •ë‹µ í‰ê·  ì‹ ë¢°ë„: {np.mean(max_probs[correct_mask]):.3f}")
    print(f"   ì˜¤ë‹µ í‰ê·  ì‹ ë¢°ë„: {np.mean(max_probs[~correct_mask]):.3f}")

    # ì‹ ë¢°ë„ë³„ ì„±ëŠ¥
    high_conf_mask = max_probs > 0.9
    medium_conf_mask = (max_probs > 0.7) & (max_probs <= 0.9)
    low_conf_mask = max_probs <= 0.7

    if np.sum(high_conf_mask) > 0:
        high_conf_acc = 100 * np.mean(correct_mask[high_conf_mask])
        print(f"   ë†’ì€ ì‹ ë¢°ë„ (>0.9): {np.sum(high_conf_mask)}ê°œ ìƒ˜í”Œ, ì •í™•ë„: {high_conf_acc:.1f}%")

    if np.sum(medium_conf_mask) > 0:
        medium_conf_acc = 100 * np.mean(correct_mask[medium_conf_mask])
        print(f"   ì¤‘ê°„ ì‹ ë¢°ë„ (0.7-0.9): {np.sum(medium_conf_mask)}ê°œ ìƒ˜í”Œ, ì •í™•ë„: {medium_conf_acc:.1f}%")

    if np.sum(low_conf_mask) > 0:
        low_conf_acc = 100 * np.mean(correct_mask[low_conf_mask])
        print(f"   ë‚®ì€ ì‹ ë¢°ë„ (<0.7): {np.sum(low_conf_mask)}ê°œ ìƒ˜í”Œ, ì •í™•ë„: {low_conf_acc:.1f}%")

    return final_accuracy, precision, recall, f1

# 10. ì„±ëŠ¥ í–¥ìƒ ì œì•ˆ í•¨ìˆ˜
def suggest_performance_improvements(accuracy, gpu_memory):
    """í˜„ì¬ ì„±ëŠ¥ê³¼ GPU ë©”ëª¨ë¦¬ ê¸°ì¤€ ê°œì„  ì œì•ˆ"""

    print(f"\n" + "=" * 80)
    print(f"ğŸ’¡ ì„±ëŠ¥ í–¥ìƒ ì œì•ˆ (í˜„ì¬: {accuracy:.2f}%, GPU: {gpu_memory:.1f}GB)")
    print("=" * 80)

    suggestions = []

    # GPU ë©”ëª¨ë¦¬ ê¸°ë°˜ ì œì•ˆ
    if gpu_memory < 8:
        suggestions.extend([
            "ğŸ”§ ë” í° GPUë¡œ ì—…ê·¸ë ˆì´ë“œ (RTX 3080/4070 ì´ìƒ)",
            "âš¡ Gradient Accumulation Steps ì¦ê°€ë¡œ ê°€ìƒ ë°°ì¹˜ í¬ê¸° í™•ëŒ€",
            "ğŸ“Š ì™¸ë¶€ ë°ì´í„°ì…‹ìœ¼ë¡œ ì‚¬ì „ í›ˆë ¨ í›„ Fine-tuning"
        ])
    elif gpu_memory >= 12:
        suggestions.extend([
            "ğŸš€ ë” í° ëª¨ë¸ ì‚¬ìš© (EfficientNet-B6, ConvNeXt-Base)",
            "ğŸ“ˆ ë” ë†’ì€ í•´ìƒë„ (384x384 â†’ 512x512)",
            "ğŸ”¥ ì•™ìƒë¸” ëª¨ë¸ êµ¬ì„± (2-3ê°œ ëª¨ë¸ ì¡°í•©)"
        ])

    # ì„±ëŠ¥ ê¸°ë°˜ ì œì•ˆ
    if accuracy < 95:
        suggestions.extend([
            "ğŸ¯ ë” ê°•í•œ ë°ì´í„° ì¦ê°• (AutoAugment, RandAugment)",
            "ğŸ”„ Self-supervised pretraining (SimCLR, SwAV)",
            "ğŸ“š Knowledge Distillation ì ìš©",
            "ğŸ² Stochastic Weight Averaging (SWA)"
        ])

    if accuracy < 97:
        suggestions.extend([
            "ğŸŒŠ Progressive Resizing í•™ìŠµë²•",
            "ğŸ¨ StyleGANì„ ì´ìš©í•œ synthetic ë°ì´í„° ìƒì„±",
            "ğŸ” Hard Negative Mining",
            "âš™ï¸ Hyperparameter Optimization (Optuna)"
        ])

    if accuracy < 98:
        suggestions.extend([
            "ğŸ§  Vision Transformer (ViT) ëª¨ë¸ ì‹œë„",
            "ğŸŒ ëŒ€ê·œëª¨ ì™¸ë¶€ ë°ì´í„°ì…‹ í™œìš©",
            "ğŸ”¬ Attention mechanism ê°•í™”",
            "ğŸª Meta-learning ê¸°ë²• ì ìš©"
        ])

    # ë©”ëª¨ë¦¬ ìµœì í™” ì¶”ê°€ ì œì•ˆ
    memory_suggestions = [
        "ğŸ’¾ Gradient Checkpointing ë” ì„¸ë°€í•˜ê²Œ ì ìš©",
        "âš¡ Mixed Precision Training ìµœì í™”",
        "ğŸ”„ Dynamic Loss Scaling ì¡°ì •",
        "ğŸ“¦ Model Parallelism (ì—¬ëŸ¬ GPU ì‚¬ìš©)",
    ]

    print("ğŸ¯ ì„±ëŠ¥ í–¥ìƒ ë°©ë²•:")
    for i, suggestion in enumerate(suggestions[:6], 1):
        print(f"   {i:2d}. {suggestion}")

    print("\nğŸ’¾ ë©”ëª¨ë¦¬ ìµœì í™” ì¶”ê°€ ë°©ë²•:")
    for i, suggestion in enumerate(memory_suggestions, 1):
        print(f"   {i:2d}. {suggestion}")

    target_acc = min(99, accuracy + 2)
    print(f"\nğŸš€ ë‹¤ìŒ ëª©í‘œ: {target_acc:.0f}%+ ë‹¬ì„±!")

# 11. ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜
def main():
    """ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜"""

    print("ğŸš€ ì§„ì§œ ë©”ëª¨ë¦¬ ìµœì í™”ëœ ê³ ì„±ëŠ¥ ì£¼ì–¼ë¦¬ ë¶„ë¥˜ ëª¨ë¸")
    print("=" * 80)
    print("ğŸ’¾ GPU ë©”ëª¨ë¦¬ ë¶€ì¡± ë¬¸ì œ ì™„ì „ í•´ê²° + ìµœê³  ì„±ëŠ¥ ë‹¬ì„±")
    print("=" * 80)

    # ë°ì´í„° ê²½ë¡œ ì„¤ì •
    data_path = '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/accesary_datasets'

    if not os.path.exists(data_path):
        print(f"âŒ ë°ì´í„° ê²½ë¡œê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {data_path}")
        print("ğŸ’¡ ë°ì´í„° ê²½ë¡œë¥¼ í™•ì¸í•˜ê±°ë‚˜ ë°ì´í„°ë¥¼ ì—…ë¡œë“œí•´ì£¼ì„¸ìš”.")
        return

    try:
        # 1ë‹¨ê³„: ë©”ëª¨ë¦¬ ìµœì í™” í•™ìŠµ
        print("\nğŸ¯ 1ë‹¨ê³„: ë©”ëª¨ë¦¬ ìµœì í™” í•™ìŠµ ì‹¤í–‰")
        model, train_losses, train_accs, val_losses, val_accs, best_acc, class_names = run_memory_optimized_training(
            data_path=data_path,
            num_epochs=50
        )

        print(f"\nâœ… 1ë‹¨ê³„ ì™„ë£Œ! ìµœê³  ì„±ëŠ¥: {best_acc:.2f}%")

        # 2ë‹¨ê³„: TTA ì ìš© ìµœì¢… í‰ê°€
        print("\nğŸ” 2ë‹¨ê³„: TTA ì ìš© ìµœì¢… í‰ê°€")

        # ê²€ì¦ ë°ì´í„°ë¡œë” ì¬ìƒì„± (TTAìš©)
        gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1024**3
        config = AdaptiveConfig(gpu_memory)

        val_transform = MemoryOptimizedTransforms.get_val_transform(config.image_size)
        val_dataset = ImageFolder(root=os.path.join(data_path, 'val'), transform=val_transform)
        val_loader = DataLoader(val_dataset, batch_size=config.batch_size, shuffle=False, num_workers=1)

        # ìµœê³  ëª¨ë¸ ë¡œë“œ
        model.load_state_dict(torch.load('/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/memory_optimized_best_model.pth'))
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

        final_accuracy, precision, recall, f1 = evaluate_with_tta(
            model, val_loader, class_names, device, num_tta=5
        )

        # 3ë‹¨ê³„: ìµœì¢… ê²°ê³¼ ë° ê°œì„  ì œì•ˆ
        print("\nğŸ’¡ 3ë‹¨ê³„: ì„±ëŠ¥ ë¶„ì„ ë° ê°œì„  ì œì•ˆ")
        suggest_performance_improvements(final_accuracy, gpu_memory)

        # ìµœì¢… ìš”ì•½
        print(f"\n" + "ğŸ‰" * 20)
        print(f"ğŸ† ìµœì¢… ê²°ê³¼ ìš”ì•½")
        print(f"ğŸ‰" * 20)
        print(f"ğŸ“ˆ ë‹¬ì„± ì„±ëŠ¥: {final_accuracy:.2f}%")
        print(f"ğŸ’¾ ìµœëŒ€ GPU ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: ~{gpu_memory * 0.6:.1f}GB (ì ˆì•½ë¨!)")
        print(f"âš¡ ì‚¬ìš©ëœ ëª¨ë¸ í¬ê¸°: {config.model_size}")
        print(f"ğŸ–¼ï¸  ì´ë¯¸ì§€ í•´ìƒë„: {config.image_size}x{config.image_size}")
        print(f"ğŸ“¦ ì‹¤ì œ ë°°ì¹˜ í¬ê¸°: {config.effective_batch_size}")
        print(f"ğŸ’¾ ëª¨ë¸ ì €ì¥ ìœ„ì¹˜: memory_optimized_best_model.pth")

        # ìµœì¢… ì‹œê°í™”
        plot_training_progress(train_losses, train_accs, val_losses, val_accs, len(train_losses))

    except RuntimeError as e:
        if "out of memory" in str(e).lower():
            print("âŒ GPU ë©”ëª¨ë¦¬ ë¶€ì¡± ì˜¤ë¥˜ ë°œìƒ")
            print("ğŸ’¡ ì¶”ê°€ ìµœì í™” ì œì•ˆ:")
            print("   1. ë” ì‘ì€ ë°°ì¹˜ í¬ê¸°ë¡œ ì¬ì‹œë„")
            print("   2. ì´ë¯¸ì§€ í•´ìƒë„ ê°ì†Œ")
            print("   3. ë” ì‘ì€ ëª¨ë¸ ì‚¬ìš©")
            print("   4. Gradient Accumulation Steps ì¦ê°€")
        else:
            print(f"âŒ ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜: {str(e)}")

        # ë©”ëª¨ë¦¬ ì •ë¦¬
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        gc.collect()

    finally:
        # ìµœì¢… ë©”ëª¨ë¦¬ ì •ë¦¬
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        gc.collect()
        print("\nğŸ§¹ ë©”ëª¨ë¦¬ ì •ë¦¬ ì™„ë£Œ")

# 12. ì¶”ê°€ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜ë“¤
def check_gpu_status():
    """GPU ìƒíƒœ í™•ì¸"""
    if torch.cuda.is_available():
        print(f"ğŸ” GPU ìƒíƒœ í™•ì¸:")
        print(f"   GPU ì´ë¦„: {torch.cuda.get_device_name()}")
        print(f"   ì´ ë©”ëª¨ë¦¬: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
        print(f"   í˜„ì¬ ì‚¬ìš©ëŸ‰: {torch.cuda.memory_allocated() / 1024**3:.1f} GB")
        print(f"   ì˜ˆì•½ ë©”ëª¨ë¦¬: {torch.cuda.memory_reserved() / 1024**3:.1f} GB")
    else:
        print("âŒ CUDAë¥¼ ì‚¬ìš©í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

def predict_single_image(model, image_path, transform, class_names, device):
    """ë‹¨ì¼ ì´ë¯¸ì§€ ì˜ˆì¸¡"""
    model.eval()

    try:
        # ì´ë¯¸ì§€ ë¡œë“œ ë° ì „ì²˜ë¦¬
        image = Image.open(image_path).convert('RGB')
        image_tensor = transform(image).unsqueeze(0).to(device)

        with torch.no_grad():
            with autocast():
                outputs = model(image_tensor)
                probabilities = F.softmax(outputs, dim=1)
                predicted_class_idx = torch.argmax(probabilities, dim=1).item()
                confidence = probabilities[0][predicted_class_idx].item()

        predicted_class = class_names[predicted_class_idx]

        print(f"ğŸ” ì˜ˆì¸¡ ê²°ê³¼:")
        print(f"   ì´ë¯¸ì§€: {os.path.basename(image_path)}")
        print(f"   ì˜ˆì¸¡ í´ë˜ìŠ¤: {predicted_class}")
        print(f"   ì‹ ë¢°ë„: {confidence:.3f}")

        # ëª¨ë“  í´ë˜ìŠ¤ë³„ í™•ë¥  ì¶œë ¥
        print(f"\nğŸ“Š ëª¨ë“  í´ë˜ìŠ¤ í™•ë¥ :")
        for i, class_name in enumerate(class_names):
            prob = probabilities[0][i].item()
            print(f"   {class_name}: {prob:.3f}")

        return predicted_class, confidence

    except Exception as e:
        print(f"âŒ ì˜ˆì¸¡ ì˜¤ë¥˜: {str(e)}")
        return None, 0.0

# ì‹¤í–‰
if __name__ == "__main__":
    # GPU ìƒíƒœ í™•ì¸
    check_gpu_status()

    # ë©”ì¸ ì‹¤í–‰
    main()
```

```python
# ì €ì¥ëœ ëª¨ë¸ í…ŒìŠ¤íŠ¸ ë° ê²°ê³¼ í™•ì¸

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.cuda.amp import autocast
from torchvision import transforms
from torchvision.datasets import ImageFolder
import timm
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import random
from pathlib import Path
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

# 1. ëª¨ë¸ í´ë˜ìŠ¤ ì •ì˜ (ì €ì¥ëœ ëª¨ë¸ê³¼ ë™ì¼í•œ êµ¬ì¡°)
class MemoryEfficientClassifier(nn.Module):
    def __init__(self, num_classes=4, model_size='xlarge'):
        super(MemoryEfficientClassifier, self).__init__()

        # ëª¨ë¸ í¬ê¸°ë³„ ë°±ë³¸ ì„ íƒ (ì €ì¥ëœ ëª¨ë¸ê³¼ ë™ì¼)
        model_configs = {
            'small': ['mobilenetv3_large_100.ra_in1k', 'efficientnet_b2.ra_in1k'],
            'medium': ['efficientnet_b3.ra2_in1k', 'efficientnet_b4.ra2_in1k'],
            'large': ['efficientnetv2_rw_m.agc_in1k', 'efficientnet_b5.ra_in1k'],
            'xlarge': [ 'efficientnetv2_m.in21k_ft_in1k', 'convnext_base.fb_in22k_ft_in1k', 'efficientnet_b6.ra_in1k' ]
        }

        # ë°±ë³¸ ë¡œë“œ
        self.backbone = None
        backbone_features = None

        for model_name in model_configs[model_size]:
            try:
                self.backbone = timm.create_model(model_name, pretrained=True)
                backbone_features = self.backbone.num_features
                self.backbone.classifier = nn.Identity()
                print(f"âœ… ëª¨ë¸ ë¡œë“œ: {model_name}")
                break
            except:
                continue

        if self.backbone is None:
            raise RuntimeError("ëª¨ë¸ì„ ë¡œë“œí•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        # ë¶„ë¥˜ê¸° í—¤ë“œ (ì €ì¥ëœ ëª¨ë¸ê³¼ ë™ì¼í•œ êµ¬ì¡°)
        if model_size == 'small':
            hidden_size = 256
        elif model_size == 'medium':
            hidden_size = 512
        elif model_size == 'large':
            hidden_size = 768
        else:  # xlarge
            hidden_size = 1024

        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(backbone_features, hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.ReLU(inplace=True),
            nn.Dropout(0.2),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.BatchNorm1d(hidden_size // 2),
            nn.ReLU(inplace=True),
            nn.Dropout(0.1),
            nn.Linear(hidden_size // 2, num_classes)
        )

    def forward(self, x):
        features = self.backbone.forward_features(x)
        output = self.classifier(features)
        return output

# 2. ë°ì´í„° ë³€í™˜ (ëª¨ë¸ í›ˆë ¨ ì‹œì™€ ë™ì¼)
def get_test_transform(image_size=320):
    return transforms.Compose([
        transforms.Resize((image_size, image_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

# 3. ê²€ì¦ í´ë”ì—ì„œ ê° í´ë˜ìŠ¤ë³„ ìƒ˜í”Œ ì„ íƒ
def select_test_samples(data_path, samples_per_class=5):
    """ê° í´ë˜ìŠ¤ë³„ë¡œ ì§€ì •ëœ ê°œìˆ˜ë§Œí¼ ìƒ˜í”Œ ì„ íƒ"""

    val_path = os.path.join(data_path, 'val')
    class_samples = {}

    if not os.path.exists(val_path):
        print(f"âŒ ê²€ì¦ í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {val_path}")
        return {}

    # í´ë˜ìŠ¤ í´ë”ë“¤ í™•ì¸
    class_dirs = [d for d in os.listdir(val_path) if os.path.isdir(os.path.join(val_path, d))]
    class_dirs.sort()

    print(f"ğŸ“ ë°œê²¬ëœ í´ë˜ìŠ¤: {class_dirs}")

    for class_name in class_dirs:
        class_path = os.path.join(val_path, class_name)

        # ì´ë¯¸ì§€ íŒŒì¼ë“¤ ì°¾ê¸°
        image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp']
        image_files = []

        for ext in image_extensions:
            image_files.extend([f for f in os.listdir(class_path)
                              if f.lower().endswith(ext.lower())])

        if image_files:
            # ëœë¤í•˜ê²Œ ìƒ˜í”Œ ì„ íƒ
            selected_samples = random.sample(image_files,
                                           min(samples_per_class, len(image_files)))

            class_samples[class_name] = [
                os.path.join(class_path, sample) for sample in selected_samples
            ]

            print(f"   {class_name}: {len(image_files)}ê°œ ì¤‘ {len(selected_samples)}ê°œ ì„ íƒ")
        else:
            print(f"   {class_name}: ì´ë¯¸ì§€ íŒŒì¼ ì—†ìŒ")

    return class_samples, class_dirs

# 4. ë‹¨ì¼ ì´ë¯¸ì§€ ì˜ˆì¸¡ (TTA ì ìš©)
def predict_single_image_with_tta(model, image_path, transform, device, num_tta=5):
    """TTAë¥¼ ì ìš©í•œ ë‹¨ì¼ ì´ë¯¸ì§€ ì˜ˆì¸¡"""

    try:
        # ì´ë¯¸ì§€ ë¡œë“œ
        image = Image.open(image_path).convert('RGB')
        image_tensor = transform(image).unsqueeze(0).to(device)

        model.eval()
        predictions = []

        with torch.no_grad():
            # ì›ë³¸ ì´ë¯¸ì§€
            with autocast():
                outputs = model(image_tensor)
                predictions.append(F.softmax(outputs, dim=1))

            # TTA ë³€í™˜ë“¤
            tta_transforms = [
                lambda x: torch.flip(x, dims=[3]),  # ìˆ˜í‰ ë°˜ì „
                lambda x: torch.flip(x, dims=[2]),  # ìˆ˜ì§ ë°˜ì „
                lambda x: torch.rot90(x, k=2, dims=[2, 3]),  # 180ë„ íšŒì „
                lambda x: x * 0.95 + 0.02,  # ë°ê¸° ì¡°ì •
            ]

            for i, tta_transform in enumerate(tta_transforms[:num_tta-1]):
                try:
                    transformed = tta_transform(image_tensor.clone())
                    with autocast():
                        outputs = model(transformed)
                        predictions.append(F.softmax(outputs, dim=1))
                except:
                    continue

            # í‰ê·  ì˜ˆì¸¡
            if predictions:
                avg_prediction = torch.mean(torch.stack(predictions), dim=0)
                predicted_class_idx = torch.argmax(avg_prediction, dim=1).item()
                confidence = avg_prediction[0][predicted_class_idx].item()
                class_probabilities = avg_prediction[0].cpu().numpy()

                return predicted_class_idx, confidence, class_probabilities

    except Exception as e:
        print(f"âŒ ì˜ˆì¸¡ ì˜¤ë¥˜ ({os.path.basename(image_path)}): {str(e)}")

    return None, 0.0, None

# 5. ë°°ì¹˜ í…ŒìŠ¤íŠ¸ ë° ê²°ê³¼ ì‹œê°í™”
def test_model_comprehensive(model_path, data_path, samples_per_class=5):
    """ëª¨ë¸ ì¢…í•© í…ŒìŠ¤íŠ¸"""

    print("ğŸ” ì €ì¥ëœ ëª¨ë¸ ì¢…í•© í…ŒìŠ¤íŠ¸ ì‹œì‘")
    print("=" * 60)

    # GPU ì„¤ì •
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"ğŸš€ ì‚¬ìš© ë””ë°”ì´ìŠ¤: {device}")

    # í…ŒìŠ¤íŠ¸ ìƒ˜í”Œ ì„ íƒ
    print(f"\nğŸ“ í…ŒìŠ¤íŠ¸ ìƒ˜í”Œ ì„ íƒ ì¤‘...")
    class_samples, class_names = select_test_samples(data_path, samples_per_class)

    if not class_samples:
        print("âŒ í…ŒìŠ¤íŠ¸í•  ìƒ˜í”Œì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return

    # ëª¨ë¸ ë¡œë“œ
    print(f"\nğŸ¤– ëª¨ë¸ ë¡œë”© ì¤‘...")
    try:
        model = MemoryEfficientClassifier(num_classes=len(class_names), model_size='xlarge')
        model.load_state_dict(torch.load(model_path, map_location=device))
        model = model.to(device)
        model.eval()
        print(f"âœ… ëª¨ë¸ ë¡œë“œ ì„±ê³µ: {model_path}")
    except Exception as e:
        print(f"âŒ ëª¨ë¸ ë¡œë“œ ì‹¤íŒ¨: {str(e)}")
        return

    # ë°ì´í„° ë³€í™˜
    transform = get_test_transform(320)

    # í…ŒìŠ¤íŠ¸ ì‹¤í–‰
    print(f"\nğŸ¯ ì˜ˆì¸¡ í…ŒìŠ¤íŠ¸ ì‹¤í–‰ ì¤‘...")
    print("=" * 80)

    all_predictions = []
    all_true_labels = []
    detailed_results = []

    for true_class_idx, (class_name, image_paths) in enumerate(class_samples.items()):
        print(f"\nğŸ“‚ {class_name.upper()} í´ë˜ìŠ¤ í…ŒìŠ¤íŠ¸:")
        print("-" * 40)

        for i, image_path in enumerate(image_paths):
            pred_class_idx, confidence, class_probs = predict_single_image_with_tta(
                model, image_path, transform, device
            )

            if pred_class_idx is not None:
                pred_class_name = class_names[pred_class_idx]
                is_correct = pred_class_idx == true_class_idx

                # ê²°ê³¼ ì €ì¥
                all_predictions.append(pred_class_idx)
                all_true_labels.append(true_class_idx)

                detailed_results.append({
                    'image': os.path.basename(image_path),
                    'true_class': class_name,
                    'pred_class': pred_class_name,
                    'confidence': confidence,
                    'correct': is_correct,
                    'class_probs': class_probs
                })

                # ê²°ê³¼ ì¶œë ¥
                status = "âœ… ì •ë‹µ" if is_correct else "âŒ ì˜¤ë‹µ"
                print(f"   {i+1}. {os.path.basename(image_path)}")
                print(f"      ì˜ˆì¸¡: {pred_class_name} (ì‹ ë¢°ë„: {confidence:.3f}) {status}")

                # ìƒìœ„ 3ê°œ ì˜ˆì¸¡ í™•ë¥ 
                top3_indices = np.argsort(class_probs)[-3:][::-1]
                print(f"      ìƒìœ„3: ", end="")
                for idx in top3_indices:
                    print(f"{class_names[idx]}({class_probs[idx]:.2f}) ", end="")
                print()

    # 6. ê²°ê³¼ ë¶„ì„ ë° ì‹œê°í™”
    print(f"\nğŸ“Š í…ŒìŠ¤íŠ¸ ê²°ê³¼ ë¶„ì„")
    print("=" * 60)

    if all_predictions and all_true_labels:
        # ì „ì²´ ì •í™•ë„
        overall_accuracy = np.mean(np.array(all_predictions) == np.array(all_true_labels))
        print(f"ğŸ¯ ì „ì²´ ì •í™•ë„: {overall_accuracy:.1%}")

        # í´ë˜ìŠ¤ë³„ ì •í™•ë„
        class_accuracies = {}
        for class_idx, class_name in enumerate(class_names):
            class_mask = np.array(all_true_labels) == class_idx
            if np.sum(class_mask) > 0:
                class_acc = np.mean(np.array(all_predictions)[class_mask] == class_idx)
                class_accuracies[class_name] = class_acc
                print(f"   {class_name}: {class_acc:.1%}")

        # í˜¼ë™ í–‰ë ¬
        cm = confusion_matrix(all_true_labels, all_predictions)

        plt.figure(figsize=(12, 10))

        # í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
        plt.subplot(2, 2, 1)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=class_names, yticklabels=class_names,
                    cbar_kws={'label': 'Count'})
        plt.title(f'ğŸ¯ Confusion Matrix\nOverall Accuracy: {overall_accuracy:.1%}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')

        # í´ë˜ìŠ¤ë³„ ì •í™•ë„ ë§‰ëŒ€ ê·¸ë˜í”„
        plt.subplot(2, 2, 2)
        classes = list(class_accuracies.keys())
        accuracies = list(class_accuracies.values())
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

        bars = plt.bar(classes, accuracies, color=colors[:len(classes)], alpha=0.8)
        plt.title('ğŸ“ˆ Class-wise Accuracy')
        plt.ylabel('Accuracy')
        plt.ylim(0, 1.1)

        # ê°’ í‘œì‹œ
        for bar, acc in zip(bars, accuracies):
            plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
                    f'{acc:.1%}', ha='center', va='bottom', fontweight='bold')

        plt.xticks(rotation=45)

        # ì‹ ë¢°ë„ ë¶„í¬
        plt.subplot(2, 2, 3)
        confidences = [result['confidence'] for result in detailed_results]
        correct_confidences = [result['confidence'] for result in detailed_results if result['correct']]
        wrong_confidences = [result['confidence'] for result in detailed_results if not result['correct']]

        plt.hist(correct_confidences, bins=10, alpha=0.7, label='Correct', color='green', density=True)
        plt.hist(wrong_confidences, bins=10, alpha=0.7, label='Wrong', color='red', density=True)
        plt.title('ğŸ” Confidence Distribution')
        plt.xlabel('Confidence')
        plt.ylabel('Density')
        plt.legend()

        # ì˜ˆì¸¡ ì‹ ë¢°ë„ë³„ ì •í™•ë„
        plt.subplot(2, 2, 4)
        confidence_ranges = [(0.0, 0.7), (0.7, 0.9), (0.9, 1.0)]
        range_accuracies = []
        range_labels = []

        for low, high in confidence_ranges:
            mask = [(low <= conf < high) for conf in confidences]
            if any(mask):
                range_acc = np.mean([detailed_results[i]['correct'] for i, m in enumerate(mask) if m])
                range_accuracies.append(range_acc)
                range_labels.append(f'{low:.1f}-{high:.1f}')
            else:
                range_accuracies.append(0)
                range_labels.append(f'{low:.1f}-{high:.1f}')

        plt.bar(range_labels, range_accuracies, color=['red', 'orange', 'green'], alpha=0.7)
        plt.title('ğŸ¯ Accuracy by Confidence Range')
        plt.ylabel('Accuracy')
        plt.ylim(0, 1.1)

        for i, acc in enumerate(range_accuracies):
            plt.text(i, acc + 0.02, f'{acc:.1%}', ha='center', va='bottom', fontweight='bold')

        plt.tight_layout()
        plt.show()

        # ìƒì„¸ ê²°ê³¼ í…Œì´ë¸”
        print(f"\nğŸ“‹ ìƒì„¸ ì˜ˆì¸¡ ê²°ê³¼:")
        print("-" * 80)
        print(f"{'ì´ë¯¸ì§€':<20} {'ì‹¤ì œ':<12} {'ì˜ˆì¸¡':<12} {'ì‹ ë¢°ë„':<8} {'ê²°ê³¼':<6}")
        print("-" * 80)

        for result in detailed_results:
            status = "âœ… ë§ìŒ" if result['correct'] else "âŒ í‹€ë¦¼"
            print(f"{result['image']:<20} {result['true_class']:<12} {result['pred_class']:<12} "
                  f"{result['confidence']:<8.3f} {status:<6}")

        # ì˜¤ë‹µ ë¶„ì„
        wrong_results = [r for r in detailed_results if not r['correct']]
        if wrong_results:
            print(f"\nğŸ” ì˜¤ë‹µ ë¶„ì„ ({len(wrong_results)}ê°œ):")
            print("-" * 50)
            for result in wrong_results:
                print(f"   {result['image']}: {result['true_class']} â†’ {result['pred_class']} "
                      f"(ì‹ ë¢°ë„: {result['confidence']:.3f})")

        # ì„±ëŠ¥ ìš”ì•½
        print(f"\nğŸ† ì„±ëŠ¥ ìš”ì•½:")
        print(f"   ì „ì²´ í…ŒìŠ¤íŠ¸ ìƒ˜í”Œ: {len(detailed_results)}ê°œ")
        print(f"   ì •ë‹µ: {len([r for r in detailed_results if r['correct']])}ê°œ")
        print(f"   ì˜¤ë‹µ: {len(wrong_results)}ê°œ")
        print(f"   ì „ì²´ ì •í™•ë„: {overall_accuracy:.1%}")
        print(f"   í‰ê·  ì‹ ë¢°ë„: {np.mean(confidences):.3f}")
        print(f"   ì •ë‹µ í‰ê·  ì‹ ë¢°ë„: {np.mean(correct_confidences):.3f}")
        if wrong_confidences:
            print(f"   ì˜¤ë‹µ í‰ê·  ì‹ ë¢°ë„: {np.mean(wrong_confidences):.3f}")

        return overall_accuracy, class_accuracies, detailed_results

    return None, None, None

# 7. ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜
def main():
    """ë©”ì¸ í…ŒìŠ¤íŠ¸ ì‹¤í–‰"""

    print("ğŸ” ì €ì¥ëœ ëª¨ë¸ í…ŒìŠ¤íŠ¸")
    print("=" * 60)

    # ê²½ë¡œ ì„¤ì •
    model_path = '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/memory_optimized_best_model.pth'
    data_path = '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/accesary_datasets'

    # ëª¨ë¸ íŒŒì¼ ì¡´ì¬ í™•ì¸
    if not os.path.exists(model_path):
        print(f"âŒ ëª¨ë¸ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {model_path}")
        print("ğŸ’¡ ë¨¼ì € ëª¨ë¸ì„ í›ˆë ¨í•˜ê³  ì €ì¥í•´ì£¼ì„¸ìš”.")
        return

    # ë°ì´í„° ê²½ë¡œ í™•ì¸
    if not os.path.exists(data_path):
        print(f"âŒ ë°ì´í„° ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {data_path}")
        print("ğŸ’¡ ë°ì´í„° ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
        return

    # ì¢…í•© í…ŒìŠ¤íŠ¸ ì‹¤í–‰
    try:
        overall_acc, class_accs, results = test_model_comprehensive(
            model_path=model_path,
            data_path=data_path,
            samples_per_class=5  # ê° í´ë˜ìŠ¤ë³„ 5ê°œì”© í…ŒìŠ¤íŠ¸
        )

        if overall_acc is not None:
            print(f"\nğŸ‰ í…ŒìŠ¤íŠ¸ ì™„ë£Œ!")
            print(f"ğŸ“ˆ ë‹¬ì„± ì„±ëŠ¥: {overall_acc:.1%}")

            if overall_acc >= 0.95:
                print("ğŸ† ìš°ìˆ˜í•œ ì„±ëŠ¥!")
            elif overall_acc >= 0.90:
                print("ğŸ‘ ì¢‹ì€ ì„±ëŠ¥!")
            elif overall_acc >= 0.80:
                print("ğŸ‘Œ ì–‘í˜¸í•œ ì„±ëŠ¥")
            else:
                print("ğŸ“ˆ ê°œì„  ì—¬ì§€ ìˆìŒ")

    except Exception as e:
        print(f"âŒ í…ŒìŠ¤íŠ¸ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {str(e)}")

    finally:
        # ë©”ëª¨ë¦¬ ì •ë¦¬
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        print(f"\nğŸ§¹ ë©”ëª¨ë¦¬ ì •ë¦¬ ì™„ë£Œ")

# 8. ê°„ë‹¨í•œ ë‹¨ì¼ ì´ë¯¸ì§€ í…ŒìŠ¤íŠ¸ í•¨ìˆ˜
def quick_test_single_image(model_path, image_path, class_names=['bracelet', 'earring', 'necklace', 'ring']):
    """ë¹ ë¥¸ ë‹¨ì¼ ì´ë¯¸ì§€ í…ŒìŠ¤íŠ¸"""

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # ëª¨ë¸ ë¡œë“œ
    model = MemoryEfficientClassifier(num_classes=len(class_names), model_size='xlarge')
    model.load_state_dict(torch.load(model_path, map_location=device))
    model = model.to(device)

    transform = get_test_transform(320)
    pred_idx, confidence, probs = predict_single_image_with_tta(model, image_path, transform, device)

    if pred_idx is not None:
        pred_class = class_names[pred_idx]
        print(f"ğŸ” ì˜ˆì¸¡ ê²°ê³¼:")
        print(f"   ì´ë¯¸ì§€: {os.path.basename(image_path)}")
        print(f"   ì˜ˆì¸¡: {pred_class}")
        print(f"   ì‹ ë¢°ë„: {confidence:.3f}")
        print(f"   ëª¨ë“  í´ë˜ìŠ¤ í™•ë¥ :")
        for i, (class_name, prob) in enumerate(zip(class_names, probs)):
            print(f"     {class_name}: {prob:.3f}")

        return pred_class, confidence

# ì‹¤í–‰
if __name__ == "__main__":
    main()
```

```python
# ê³¼ì í•© í•´ê²°ì„ ìœ„í•œ ê°œì„ ëœ ë©”ëª¨ë¦¬ ìµœì í™” ëª¨ë¸

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, WeightedRandomSampler
from torch.cuda.amp import GradScaler, autocast
from torchvision import transforms
from torchvision.datasets import ImageFolder
import timm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from PIL import Image
import os
from tqdm import tqdm
import warnings
import gc
warnings.filterwarnings('ignore')

# 1. ê³¼ì í•© ë°©ì§€ë¥¼ ìœ„í•œ ê°•í™”ëœ ì„¤ì •
class OverfittingPreventionConfig:
    def __init__(self, gpu_memory_gb):
        self.gpu_memory = gpu_memory_gb

        # GPU ë©”ëª¨ë¦¬ë³„ ì„¤ì •ì€ ìœ ì§€í•˜ë˜, ê³¼ì í•© ë°©ì§€ ê°•í™”
        if gpu_memory_gb < 6:
            self.image_size = 224
            self.batch_size = 4
            self.accumulation_steps = 8
            self.num_workers = 1
            self.model_size = 'small'
        elif gpu_memory_gb < 10:
            self.image_size = 256
            self.batch_size = 6
            self.accumulation_steps = 6
            self.num_workers = 2
            self.model_size = 'medium'
        elif gpu_memory_gb < 14:
            self.image_size = 288
            self.batch_size = 8
            self.accumulation_steps = 4
            self.num_workers = 2
            self.model_size = 'large'
        else:
            self.image_size = 320
            self.batch_size = 12
            self.accumulation_steps = 2
            self.num_workers = 3
            self.model_size = 'xlarge'

        self.effective_batch_size = self.batch_size * self.accumulation_steps

        # ğŸ”¥ ê³¼ì í•© ë°©ì§€ ê°•í™” ì„¤ì •
        self.dropout_rate = 0.5  # ë“œë¡­ì•„ì›ƒ ê°•í™”
        self.weight_decay = 0.05  # ê°€ì¤‘ì¹˜ ê°ì‡  ê°•í™”
        self.label_smoothing = 0.2  # ë ˆì´ë¸” ìŠ¤ë¬´ë”© ê°•í™”
        self.mixup_alpha = 0.4  # Mixup ê°•í™”
        self.cutmix_alpha = 1.0  # CutMix ì¶”ê°€
        self.early_stop_patience = 10  # Early stopping ë” ë¹¨ë¦¬

        print(f"ğŸ›¡ï¸ ê³¼ì í•© ë°©ì§€ ê°•í™” ì„¤ì •:")
        print(f"   ë“œë¡­ì•„ì›ƒ: {self.dropout_rate}")
        print(f"   ê°€ì¤‘ì¹˜ ê°ì‡ : {self.weight_decay}")
        print(f"   ë ˆì´ë¸” ìŠ¤ë¬´ë”©: {self.label_smoothing}")
        print(f"   Early stopping: {self.early_stop_patience} patience")

# 2. ê³¼ì í•© ë°©ì§€ ê°•í™”ëœ ë°ì´í„° ì¦ê°•
class AntiOverfittingTransforms:
    @staticmethod
    def get_train_transform(image_size):
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.RandomResizedCrop(image_size, scale=(0.7, 1.0)),  # ë” ê°•í•œ í¬ë¡­
            transforms.RandomHorizontalFlip(p=0.6),  # í™•ë¥  ì¦ê°€
            transforms.RandomVerticalFlip(p=0.4),    # í™•ë¥  ì¦ê°€
            transforms.RandomRotation(degrees=30),   # íšŒì „ ê°ë„ ì¦ê°€
            transforms.RandomAffine(
                degrees=0,
                translate=(0.2, 0.2),     # ì´ë™ ë²”ìœ„ ì¦ê°€
                scale=(0.8, 1.2),         # ìŠ¤ì¼€ì¼ ë³€í™” ì¦ê°€
                shear=15                  # ì „ë‹¨ ë³€í™˜ ì¶”ê°€
            ),
            transforms.ColorJitter(
                brightness=0.3,    # ë°ê¸° ë³€í™” ì¦ê°€
                contrast=0.3,      # ëŒ€ë¹„ ë³€í™” ì¦ê°€
                saturation=0.3,    # ì±„ë„ ë³€í™” ì¦ê°€
                hue=0.15           # ìƒ‰ì¡° ë³€í™” ì¦ê°€
            ),
            transforms.RandomPerspective(distortion_scale=0.3, p=0.4),  # ì›ê·¼ ë³€í™˜ ì¶”ê°€
            transforms.RandomGrayscale(p=0.1),  # í‘ë°± ë³€í™˜ ì¶”ê°€
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
            transforms.RandomErasing(p=0.4, scale=(0.05, 0.3), ratio=(0.3, 3.3)),  # ë” ê°•í•œ erasing
        ])

    @staticmethod
    def get_val_transform(image_size):
        return transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

# 3. ê³¼ì í•© ë°©ì§€ ê°•í™”ëœ ëª¨ë¸
class OverfittingResistantClassifier(nn.Module):
    def __init__(self, num_classes=4, model_size='medium', dropout_rate=0.5):
        super(OverfittingResistantClassifier, self).__init__()

        # ë” ë³´ìˆ˜ì ì¸(ì‘ì€) ëª¨ë¸ ì„ íƒìœ¼ë¡œ ê³¼ì í•© ë°©ì§€
        model_configs = {
            'small': [
                'mobilenetv3_large_100.ra_in1k',
                'efficientnet_b0.ra_in1k',
                'resnet34.a1_in1k'
            ],
            'medium': [
                'efficientnet_b2.ra_in1k',    # B3 ëŒ€ì‹  B2
                'efficientnet_b3.ra2_in1k',
                'mobilevit_s.cvnets_in1k'
            ],
            'large': [
                'efficientnet_b4.ra2_in1k',   # B5 ëŒ€ì‹  B4
                'efficientnetv2_rw_s.ra2_in1k',  # ë” ì‘ì€ ë²„ì „
                'convnext_tiny.fb_in22k_ft_in1k'  # Base ëŒ€ì‹  Tiny
            ],
            'xlarge': [
                'efficientnet_b5.ra_in1k',    # B6 ëŒ€ì‹  B5
                'efficientnetv2_rw_m.agc_in1k',
                'convnext_small.fb_in22k_ft_in1k'  # Base ëŒ€ì‹  Small
            ]
        }

        # ë°±ë³¸ ë¡œë“œ (ë” ë³´ìˆ˜ì ìœ¼ë¡œ)
        self.backbone = None
        backbone_features = None

        for model_name in model_configs[model_size]:
            try:
                self.backbone = timm.create_model(model_name, pretrained=True)
                backbone_features = self.backbone.num_features
                self.backbone.classifier = nn.Identity()

                # Gradient Checkpointing
                if hasattr(self.backbone, 'set_grad_checkpointing'):
                    self.backbone.set_grad_checkpointing(True)

                print(f"âœ… ê³¼ì í•© ë°©ì§€ ëª¨ë¸ ë¡œë“œ: {model_name}")
                break
            except Exception as e:
                continue

        if self.backbone is None:
            raise RuntimeError("ì‚¬ìš© ê°€ëŠ¥í•œ ëª¨ë¸ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

        # ğŸ›¡ï¸ ê³¼ì í•© ë°©ì§€ ê°•í™”ëœ ë¶„ë¥˜ê¸° í—¤ë“œ
        if model_size == 'small':
            hidden_size = 128  # ë” ì‘ê²Œ
        elif model_size == 'medium':
            hidden_size = 256  # ë” ì‘ê²Œ
        elif model_size == 'large':
            hidden_size = 384  # ë” ì‘ê²Œ
        else:  # xlarge
            hidden_size = 512  # ë” ì‘ê²Œ

        # ë” ê°„ë‹¨í•œ ë¶„ë¥˜ê¸° (ê³¼ì í•© ë°©ì§€)
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),

            # ì²« ë²ˆì§¸ ë ˆì´ì–´
            nn.Dropout(dropout_rate),
            nn.Linear(backbone_features, hidden_size),
            nn.BatchNorm1d(hidden_size),
            nn.ReLU(inplace=True),

            # ë‘ ë²ˆì§¸ ë ˆì´ì–´ (ì„ íƒì )
            nn.Dropout(dropout_rate * 0.7),
            nn.Linear(hidden_size, hidden_size // 2),
            nn.BatchNorm1d(hidden_size // 2),
            nn.ReLU(inplace=True),

            # ì¶œë ¥ ë ˆì´ì–´
            nn.Dropout(dropout_rate * 0.5),
            nn.Linear(hidden_size // 2, num_classes)
        )

        # ê°€ì¤‘ì¹˜ ì´ˆê¸°í™” (He ì´ˆê¸°í™”)
        self._initialize_weights()

        total_params = sum(p.numel() for p in self.parameters())
        trainable_params = sum(p.numel() for p in self.parameters() if p.requires_grad)
        print(f"ğŸ“Š ê³¼ì í•© ë°©ì§€ ëª¨ë¸ íŒŒë¼ë¯¸í„°: {total_params:,} (í•™ìŠµ: {trainable_params:,})")

    def _initialize_weights(self):
        for m in self.classifier.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm1d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        features = self.backbone.forward_features(x)
        output = self.classifier(features)
        return output

# 4. ê³¼ì í•© ë°©ì§€ íŠ¸ë ˆì´ë„ˆ
class AntiOverfittingTrainer:
    def __init__(self, model, criterion, optimizer, scheduler, device, config):
        self.model = model
        self.criterion = criterion
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device
        self.config = config
        self.scaler = GradScaler()

        print(f"ğŸ›¡ï¸ ê³¼ì í•© ë°©ì§€ íŠ¸ë ˆì´ë„ˆ ì´ˆê¸°í™”")
        print(f"   Mixup Alpha: {config.mixup_alpha}")
        print(f"   CutMix Alpha: {config.cutmix_alpha}")

    def train_epoch(self, train_loader, epoch):
        """ê³¼ì í•© ë°©ì§€ ê°•í™”ëœ í›ˆë ¨"""
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        self.optimizer.zero_grad()
        progress_bar = tqdm(train_loader, desc=f'ğŸ›¡ï¸ Anti-Overfitting Epoch {epoch}')

        for batch_idx, (images, labels) in enumerate(progress_bar):
            images = images.to(self.device, non_blocking=True)
            labels = labels.to(self.device, non_blocking=True)

            with autocast():
                # ğŸ² ë°ì´í„° ì¦ê°• ê¸°ë²• ëœë¤ ì„ íƒ (ë” ë‹¤ì–‘í•˜ê²Œ)
                aug_choice = np.random.choice(['none', 'mixup', 'cutmix'], p=[0.2, 0.4, 0.4])

                if aug_choice == 'mixup':
                    images, labels_a, labels_b, lam = self._mixup_data(images, labels, self.config.mixup_alpha)
                    outputs = self.model(images)
                    loss = lam * self.criterion(outputs, labels_a) + (1 - lam) * self.criterion(outputs, labels_b)
                elif aug_choice == 'cutmix':
                    images, labels_a, labels_b, lam = self._cutmix_data(images, labels, self.config.cutmix_alpha)
                    outputs = self.model(images)
                    loss = lam * self.criterion(outputs, labels_a) + (1 - lam) * self.criterion(outputs, labels_b)
                else:
                    outputs = self.model(images)
                    loss = self.criterion(outputs, labels)

                # Gradient Accumulation
                loss = loss / self.config.accumulation_steps

            # Backward
            self.scaler.scale(loss).backward()

            # í†µê³„ ì—…ë°ì´íŠ¸
            running_loss += loss.item() * self.config.accumulation_steps
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)

            # ì¦ê°•ëœ ê²½ìš° ê·¼ì‚¬ ì •í™•ë„
            if aug_choice in ['mixup', 'cutmix']:
                correct += (predicted == labels).sum().item() * 0.6  # ë” ë³´ìˆ˜ì 
            else:
                correct += (predicted == labels).sum().item()

            # Gradient step
            if (batch_idx + 1) % self.config.accumulation_steps == 0:
                # ğŸ”¥ ë” ê°•í•œ Gradient Clipping
                self.scaler.unscale_(self.optimizer)
                torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=0.5)

                self.scaler.step(self.optimizer)
                self.scaler.update()
                self.optimizer.zero_grad()

                # ë©”ëª¨ë¦¬ ì •ë¦¬
                if (batch_idx + 1) % (self.config.accumulation_steps * 15) == 0:
                    torch.cuda.empty_cache()

            # ì§„í–‰ë¥  ì—…ë°ì´íŠ¸
            if batch_idx % 10 == 0:
                current_acc = 100. * correct / total if total > 0 else 0
                allocated = torch.cuda.memory_allocated() / 1024**3
                progress_bar.set_postfix({
                    'Loss': f'{loss.item() * self.config.accumulation_steps:.4f}',
                    'Acc': f'{current_acc:.2f}%',
                    'Aug': aug_choice,
                    'GPU': f'{allocated:.1f}GB'
                })

            del images, labels, outputs, loss

        # ë§ˆì§€ë§‰ ë°°ì¹˜ ì²˜ë¦¬
        if (batch_idx + 1) % self.config.accumulation_steps != 0:
            self.scaler.unscale_(self.optimizer)
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=0.5)
            self.scaler.step(self.optimizer)
            self.scaler.update()
            self.optimizer.zero_grad()

        self.scheduler.step()
        torch.cuda.empty_cache()
        gc.collect()

        epoch_loss = running_loss / len(train_loader)
        epoch_acc = 100. * correct / total

        return epoch_loss, epoch_acc

    def validate_epoch(self, val_loader):
        """ê²€ì¦"""
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, labels in tqdm(val_loader, desc='ğŸ” Validation'):
                images = images.to(self.device, non_blocking=True)
                labels = labels.to(self.device, non_blocking=True)

                with autocast():
                    outputs = self.model(images)
                    loss = self.criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                del images, labels, outputs, loss

        torch.cuda.empty_cache()

        epoch_loss = running_loss / len(val_loader)
        epoch_acc = 100. * correct / total

        return epoch_loss, epoch_acc, all_preds, all_labels

    def _mixup_data(self, x, y, alpha=1.0):
        """Mixup"""
        if alpha > 0:
            lam = np.random.beta(alpha, alpha)
        else:
            lam = 1

        batch_size = x.size(0)
        index = torch.randperm(batch_size).to(x.device)
        mixed_x = lam * x + (1 - lam) * x[index, :]
        y_a, y_b = y, y[index]
        return mixed_x, y_a, y_b, lam

    def _cutmix_data(self, x, y, alpha=1.0):
        """CutMix"""
        if alpha > 0:
            lam = np.random.beta(alpha, alpha)
        else:
            lam = 1

        batch_size = x.size(0)
        index = torch.randperm(batch_size).to(x.device)

        # CutMix bbox ìƒì„±
        W, H = x.size(2), x.size(3)
        cut_rat = np.sqrt(1. - lam)
        cut_w = np.int32(W * cut_rat)
        cut_h = np.int32(H * cut_rat)

        cx = np.random.randint(W)
        cy = np.random.randint(H)

        bbx1 = np.clip(cx - cut_w // 2, 0, W)
        bby1 = np.clip(cy - cut_h // 2, 0, H)
        bbx2 = np.clip(cx + cut_w // 2, 0, W)
        bby2 = np.clip(cy + cut_h // 2, 0, H)

        x[:, :, bbx1:bbx2, bby1:bby2] = x[index, :, bbx1:bbx2, bby1:bby2]

        # adjust lambda
        lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (W * H))

        y_a, y_b = y, y[index]
        return x, y_a, y_b, lam

# 5. ê³¼ì í•© ê°ì§€ ë° ëª¨ë‹ˆí„°ë§
class OverfittingMonitor:
    def __init__(self, patience=5):
        self.patience = patience
        self.train_accs = []
        self.val_accs = []
        self.overfitting_warnings = 0

    def check_overfitting(self, train_acc, val_acc):
        """ê³¼ì í•© ê°ì§€"""
        self.train_accs.append(train_acc)
        self.val_accs.append(val_acc)

        # ìµœê·¼ 3ê°œ ì—í¬í¬ì˜ í‰ê· ìœ¼ë¡œ íŒë‹¨
        if len(self.train_accs) >= 3:
            recent_train = np.mean(self.train_accs[-3:])
            recent_val = np.mean(self.val_accs[-3:])

            # ğŸš¨ ê³¼ì í•© ê°ì§€ ì¡°ê±´
            conditions = [
                val_acc > train_acc + 5,      # ê²€ì¦ì´ í›ˆë ¨ë³´ë‹¤ 5% ì´ìƒ ë†’ìŒ
                recent_val > recent_train + 3, # ìµœê·¼ í‰ê· ë„ 3% ì´ìƒ ì°¨ì´
                val_acc > 97 and train_acc < 85  # ê²€ì¦ì€ ë†’ì€ë° í›ˆë ¨ì€ ë‚®ìŒ
            ]

            is_overfitting = any(conditions)

            if is_overfitting:
                self.overfitting_warnings += 1
                warning_msg = f"ğŸš¨ ê³¼ì í•© ê²½ê³  #{self.overfitting_warnings}"
                warning_msg += f"\n   Train Acc: {train_acc:.2f}% vs Val Acc: {val_acc:.2f}%"
                warning_msg += f"\n   ì°¨ì´: {val_acc - train_acc:.2f}%"

                if val_acc > train_acc + 10:
                    warning_msg += "\n   âš ï¸ ë§¤ìš° ì‹¬ê°í•œ ê³¼ì í•©!"
                elif val_acc > train_acc + 7:
                    warning_msg += "\n   âš ï¸ ì‹¬ê°í•œ ê³¼ì í•©"
                else:
                    warning_msg += "\n   âš ï¸ ê°€ë²¼ìš´ ê³¼ì í•©"

                print(warning_msg)
                return True, self.overfitting_warnings

        return False, 0

# 6. ë©”ì¸ ê³¼ì í•© ë°©ì§€ í•™ìŠµ í•¨ìˆ˜
def run_anti_overfitting_training(data_path, num_epochs=40):
    """ê³¼ì í•© ë°©ì§€ ê°•í™” í•™ìŠµ"""

    print("ğŸ›¡ï¸ ê³¼ì í•© ë°©ì§€ ê°•í™” í•™ìŠµ ì‹œì‘!")
    print("=" * 80)

    # GPU ë©”ëª¨ë¦¬ ìµœì í™”
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        gc.collect()
        gpu_memory = torch.cuda.get_device_properties(0).total_memory / 1024**3
        print(f"ğŸ“Š GPU ë©”ëª¨ë¦¬: {gpu_memory:.1f} GB")
    else:
        gpu_memory = 0

    # ê³¼ì í•© ë°©ì§€ ì„¤ì •
    config = OverfittingPreventionConfig(gpu_memory)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # ê°•í™”ëœ ë°ì´í„° ë³€í™˜
    train_transform = AntiOverfittingTransforms.get_train_transform(config.image_size)
    val_transform = AntiOverfittingTransforms.get_val_transform(config.image_size)

    # ë°ì´í„° ë¡œë“œ
    train_dataset = ImageFolder(root=os.path.join(data_path, 'train'), transform=train_transform)
    val_dataset = ImageFolder(root=os.path.join(data_path, 'val'), transform=val_transform)

    print(f"ğŸ“ ë°ì´í„° ë¡œë”©:")
    print(f"   í›ˆë ¨: {len(train_dataset)}ê°œ, ê²€ì¦: {len(val_dataset)}ê°œ")
    print(f"   í´ë˜ìŠ¤: {train_dataset.classes}")

    # ê°€ì¤‘ ìƒ˜í”ŒëŸ¬
    def create_weighted_sampler(dataset):
        targets = [dataset[i][1] for i in range(len(dataset))]
        class_counts = torch.bincount(torch.tensor(targets))
        weights = 1.0 / class_counts.float()
        sample_weights = weights[targets]
        return WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

    weighted_sampler = create_weighted_sampler(train_dataset)

    # ë°ì´í„°ë¡œë”
    train_loader = DataLoader(
        train_dataset, batch_size=config.batch_size, sampler=weighted_sampler,
        num_workers=config.num_workers, pin_memory=True, persistent_workers=True if config.num_workers > 0 else False
    )
    val_loader = DataLoader(
        val_dataset, batch_size=config.batch_size * 2, shuffle=False,
        num_workers=config.num_workers, pin_memory=True, persistent_workers=True if config.num_workers > 0 else False
    )

    # ê³¼ì í•© ë°©ì§€ ëª¨ë¸
    model = OverfittingResistantClassifier(
        num_classes=len(train_dataset.classes),
        model_size=config.model_size,
        dropout_rate=config.dropout_rate
    ).to(device)

    # í´ë˜ìŠ¤ ê°€ì¤‘ì¹˜
    class_weights = compute_class_weight(
        'balanced', classes=np.unique([train_dataset[i][1] for i in range(len(train_dataset))]),
        y=[train_dataset[i][1] for i in range(len(train_dataset))]
    )
    class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

    # ê³¼ì í•© ë°©ì§€ ì†ì‹¤í•¨ìˆ˜ (ë ˆì´ë¸” ìŠ¤ë¬´ë”© ê°•í™”)
    criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=config.label_smoothing)

    # ë” ë³´ìˆ˜ì ì¸ ì˜µí‹°ë§ˆì´ì € (ê°€ì¤‘ì¹˜ ê°ì‡  ê°•í™”)
    optimizer = optim.AdamW(
        model.parameters(), lr=0.0005,  # ë” ë‚®ì€ í•™ìŠµë¥ 
        weight_decay=config.weight_decay, betas=(0.9, 0.999)
    )

    # ë” ë¶€ë“œëŸ¬ìš´ ìŠ¤ì¼€ì¤„ëŸ¬
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs, eta_min=1e-6)

    # íŠ¸ë ˆì´ë„ˆ ë° ëª¨ë‹ˆí„° ì´ˆê¸°í™”
    trainer = AntiOverfittingTrainer(model, criterion, optimizer, scheduler, device, config)
    overfitting_monitor = OverfittingMonitor(patience=3)

    # í•™ìŠµ ë³€ìˆ˜
    train_losses, train_accs = [], []
    val_losses, val_accs = [], []
    best_val_acc = 0.0
    patience_counter = 0

    print(f"\nğŸ¯ ê³¼ì í•© ë°©ì§€ í•™ìŠµ ì‹œì‘ (ì—í¬í¬: {num_epochs})")
    print("=" * 60)

    for epoch in range(num_epochs):
        print(f'\nğŸ“… Epoch {epoch+1}/{num_epochs}')
        print('-' * 50)

        # í›ˆë ¨
        train_loss, train_acc = trainer.train_epoch(train_loader, epoch)

        # ê²€ì¦
        val_loss, val_acc, val_preds, val_labels = trainer.validate_epoch(val_loader)

        # ê¸°ë¡
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        current_lr = optimizer.param_groups[0]['lr']

        print(f'ğŸ“Š Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%')
        print(f'ğŸ“Š Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%')
        print(f'ğŸ¯ Learning Rate: {current_lr:.2e}')
        print(f'ğŸ“ˆ Train vs Val Gap: {val_acc - train_acc:+.2f}%')

        # ğŸš¨ ê³¼ì í•© ëª¨ë‹ˆí„°ë§
        is_overfitting, warning_count = overfitting_monitor.check_overfitting(train_acc, val_acc)

        # ê±´ì „í•œ ê¸°ì¤€ìœ¼ë¡œ ìµœê³  ëª¨ë¸ ì €ì¥
        # ê²€ì¦ ì •í™•ë„ê°€ í›ˆë ¨ ì •í™•ë„ë³´ë‹¤ ë„ˆë¬´ ë†’ì§€ ì•Šì„ ë•Œë§Œ ì €ì¥
        healthy_gap = val_acc - train_acc
        if val_acc > best_val_acc and healthy_gap <= 8:  # 8% ì´ë‚´ ì°¨ì´
            best_val_acc = val_acc
            torch.save(model.state_dict(), 'anti_overfitting_best_model.pth')
            print(f'ğŸ‰ ê±´ì „í•œ ì„±ëŠ¥ í–¥ìƒ! ëª¨ë¸ ì €ì¥ (Val Acc: {best_val_acc:.2f}%, Gap: {healthy_gap:+.2f}%)')
            patience_counter = 0
        elif val_acc > best_val_acc:
            print(f'âš ï¸ ì„±ëŠ¥ì€ í–¥ìƒë˜ì—ˆì§€ë§Œ ê³¼ì í•© ì˜ì‹¬ (Gap: {healthy_gap:+.2f}%)')
            patience_counter += 1
        else:
            patience_counter += 1

        # Early stopping (ê³¼ì í•© ë°©ì§€ ê°•í™”)
        if patience_counter >= config.early_stop_patience:
            print(f"\nâ° Early stopping! {config.early_stop_patience} ì—í¬í¬ ë™ì•ˆ ê±´ì „í•œ ê°œì„  ì—†ìŒ")
            break

        # ê³¼ì í•©ì´ ì‹¬ê°í•´ì§€ë©´ í•™ìŠµ ì¤‘ë‹¨
        if warning_count >= 5:
            print(f"\nğŸš¨ ì‹¬ê°í•œ ê³¼ì í•©ìœ¼ë¡œ í•™ìŠµ ì¤‘ë‹¨! (ê²½ê³  {warning_count}íšŒ)")
            break

    print(f'\nğŸ›¡ï¸ ê³¼ì í•© ë°©ì§€ í•™ìŠµ ì™„ë£Œ!')
    print(f'ğŸ† ìµœê³  ê±´ì „í•œ ê²€ì¦ ì •í™•ë„: {best_val_acc:.2f}%')

    # ìµœì¢… ë©”ëª¨ë¦¬ ì •ë¦¬
    torch.cuda.empty_cache()
    gc.collect()

    return model, train_losses, train_accs, val_losses, val_accs, best_val_acc, train_dataset.classes

# 7. ê³¼ì í•© ì§„ë‹¨ ë° ì‹œê°í™”
def diagnose_overfitting(train_losses, train_accs, val_losses, val_accs):
    """ê³¼ì í•© ìƒíƒœ ì§„ë‹¨ ë° ì‹œê°í™”"""

    print("\nğŸ” ê³¼ì í•© ì§„ë‹¨ ë¶„ì„")
    print("=" * 50)

    # ìµœê·¼ 5 ì—í¬í¬ í‰ê· ìœ¼ë¡œ ë¶„ì„
    if len(train_accs) >= 5:
        recent_train = np.mean(train_accs[-5:])
        recent_val = np.mean(val_accs[-5:])
        gap = recent_val - recent_train

        print(f"ğŸ“Š ìµœê·¼ 5 ì—í¬í¬ í‰ê· :")
        print(f"   í›ˆë ¨ ì •í™•ë„: {recent_train:.2f}%")
        print(f"   ê²€ì¦ ì •í™•ë„: {recent_val:.2f}%")
        print(f"   ì„±ëŠ¥ ì°¨ì´: {gap:+.2f}%")

        # ê³¼ì í•© ìˆ˜ì¤€ íŒë‹¨
        if gap > 15:
            status = "ğŸš¨ ê·¹ì‹¬í•œ ê³¼ì í•©"
            color = 'red'
        elif gap > 10:
            status = "âš ï¸ ì‹¬í•œ ê³¼ì í•©"
            color = 'orange'
        elif gap > 5:
            status = "âš ï¸ ê°€ë²¼ìš´ ê³¼ì í•©"
            color = 'yellow'
        elif gap > -2:
            status = "âœ… ê±´ì „í•œ ìƒíƒœ"
            color = 'green'
        else:
            status = "ğŸ“‰ ì–¸ë”í”¼íŒ…"
            color = 'blue'

        print(f"ğŸ¯ ì§„ë‹¨ ê²°ê³¼: {status}")

        # ì‹œê°í™”
        fig, axes = plt.subplots(1, 3, figsize=(20, 6))

        epochs = range(1, len(train_losses) + 1)

        # 1. Loss ê·¸ë˜í”„
        axes[0].plot(epochs, train_losses, 'b-', label='Train Loss', linewidth=2)
        axes[0].plot(epochs, val_losses, 'r-', label='Validation Loss', linewidth=2)
        axes[0].fill_between(epochs, train_losses, alpha=0.3, color='blue')
        axes[0].fill_between(epochs, val_losses, alpha=0.3, color='red')
        axes[0].set_title('ğŸ“‰ Loss Curves', fontsize=14, fontweight='bold')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Loss')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)

        # 2. Accuracy ê·¸ë˜í”„
        axes[1].plot(epochs, train_accs, 'b-', label='Train Accuracy', linewidth=2)
        axes[1].plot(epochs, val_accs, 'r-', label='Validation Accuracy', linewidth=2)
        axes[1].fill_between(epochs, train_accs, alpha=0.3, color='blue')
        axes[1].fill_between(epochs, val_accs, alpha=0.3, color='red')
        axes[1].set_title('ğŸ“ˆ Accuracy Curves', fontsize=14, fontweight='bold')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Accuracy (%)')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)

        # ê³¼ì í•© ì˜ì—­ í‘œì‹œ
        if gap > 5:
            axes[1].axhspan(max(train_accs), max(val_accs), alpha=0.2, color='red',
                           label='Overfitting Zone')

        # 3. ì„±ëŠ¥ ì°¨ì´ ê·¸ë˜í”„
        performance_gaps = np.array(val_accs) - np.array(train_accs)
        axes[2].plot(epochs, performance_gaps, 'g-', linewidth=3, label='Val - Train Gap')
        axes[2].fill_between(epochs, performance_gaps, alpha=0.3, color='green')
        axes[2].axhline(y=0, color='black', linestyle='--', alpha=0.5, label='Perfect Balance')
        axes[2].axhline(y=5, color='orange', linestyle='--', alpha=0.7, label='Overfitting Threshold')
        axes[2].axhline(y=10, color='red', linestyle='--', alpha=0.7, label='Severe Overfitting')

        # í˜„ì¬ ìƒíƒœ í‘œì‹œ
        if performance_gaps[-1] > 5:
            axes[2].scatter([len(epochs)], [performance_gaps[-1]],
                          color='red', s=100, zorder=5, label='Current State')

        axes[2].set_title('ğŸ¯ Overfitting Monitor', fontsize=14, fontweight='bold')
        axes[2].set_xlabel('Epoch')
        axes[2].set_ylabel('Performance Gap (%)')
        axes[2].legend()
        axes[2].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        return status, gap

    return "ë¶„ì„ ë¶ˆê°€", 0

# 8. ê³¼ì í•© í•´ê²°ì±… ì œì•ˆ
def suggest_overfitting_solutions(gap, current_accuracy):
    """ê³¼ì í•© ìˆ˜ì¤€ì— ë”°ë¥¸ í•´ê²°ì±… ì œì•ˆ"""

    print(f"\nğŸ’¡ ê³¼ì í•© í•´ê²° ë°©ì•ˆ (ì„±ëŠ¥ ì°¨ì´: {gap:+.2f}%)")
    print("=" * 60)

    solutions = []

    if gap > 15:  # ê·¹ì‹¬í•œ ê³¼ì í•©
        solutions.extend([
            "ğŸš¨ ì¦‰ì‹œ ì¡°ì¹˜ í•„ìš”:",
            "   â€¢ ë“œë¡­ì•„ì›ƒì„ 0.7-0.8ë¡œ ëŒ€í­ ì¦ê°€",
            "   â€¢ ëª¨ë¸ í¬ê¸°ë¥¼ 50% ì´ìƒ ì¶•ì†Œ",
            "   â€¢ ê°€ì¤‘ì¹˜ ê°ì‡ ë¥¼ 0.1-0.2ë¡œ ê°•í™”",
            "   â€¢ í•™ìŠµë¥ ì„ 1/10ë¡œ ê°ì†Œ",
            "   â€¢ ë°ì´í„° ì¦ê°•ì„ 2ë°° ê°•í™”",
            "   â€¢ ì¡°ê¸° ì¢…ë£Œ patienceë¥¼ 3-5ë¡œ ì¶•ì†Œ"
        ])
    elif gap > 10:  # ì‹¬í•œ ê³¼ì í•©
        solutions.extend([
            "âš ï¸ ê°•ë ¥í•œ ì •ê·œí™” í•„ìš”:",
            "   â€¢ ë“œë¡­ì•„ì›ƒì„ 0.6ìœ¼ë¡œ ì¦ê°€",
            "   â€¢ ë ˆì´ë¸” ìŠ¤ë¬´ë”©ì„ 0.3ìœ¼ë¡œ ì¦ê°€",
            "   â€¢ Mixup/CutMix ê°•í™” (alpha=0.5)",
            "   â€¢ ë°°ì¹˜ í¬ê¸° 2ë°° ì¦ê°€",
            "   â€¢ ë” ê°•í•œ ë°ì´í„° ì¦ê°• ì ìš©"
        ])
    elif gap > 5:  # ê°€ë²¼ìš´ ê³¼ì í•©
        solutions.extend([
            "âš ï¸ ì ë‹¹í•œ ì •ê·œí™”:",
            "   â€¢ ë“œë¡­ì•„ì›ƒì„ 0.4-0.5ë¡œ ì¦ê°€",
            "   â€¢ ê°€ì¤‘ì¹˜ ê°ì‡ ë¥¼ 0.05ë¡œ ì¦ê°€",
            "   â€¢ ë°ì´í„° ì¦ê°• ë‹¤ì–‘í™”",
            "   â€¢ í•™ìŠµë¥ ì„ ì ˆë°˜ìœ¼ë¡œ ê°ì†Œ"
        ])
    else:  # ê±´ì „í•œ ìƒíƒœ
        solutions.extend([
            "âœ… í˜„ì¬ ìƒíƒœ ì–‘í˜¸:",
            "   â€¢ í˜„ì¬ ì„¤ì • ìœ ì§€ ê¶Œì¥",
            "   â€¢ í•„ìš”ì‹œ ë¯¸ì„¸ ì¡°ì •ë§Œ ì§„í–‰",
            "   â€¢ ë” ê¸´ í•™ìŠµìœ¼ë¡œ ì„±ëŠ¥ ê°œì„  ê°€ëŠ¥"
        ])

    # ì¶”ê°€ ê³ ê¸‰ ê¸°ë²•ë“¤
    advanced_solutions = [
        "ğŸ”¬ ê³ ê¸‰ ì •ê·œí™” ê¸°ë²•:",
        "   â€¢ Stochastic Depth ì ìš©",
        "   â€¢ Spectral Normalization",
        "   â€¢ Gradient Penalty ì¶”ê°€",
        "   â€¢ Knowledge Distillation",
        "   â€¢ Self-supervised Pre-training",
        "",
        "ğŸ“Š ë°ì´í„° ê´€ë ¨ í•´ê²°ì±…:",
        "   â€¢ ë” ë§ì€ í›ˆë ¨ ë°ì´í„° ìˆ˜ì§‘",
        "   â€¢ í´ë˜ìŠ¤ë³„ ë°ì´í„° ê· í˜• ë§ì¶”ê¸°",
        "   â€¢ ì™¸ë¶€ ë°ì´í„°ì…‹ í™œìš©",
        "   â€¢ Pseudo-labeling ì ìš©"
    ]

    solutions.extend(advanced_solutions)

    for solution in solutions:
        print(solution)

    # ë§ì¶¤í˜• ê¶Œì¥ì‚¬í•­
    if current_accuracy < 85:
        print(f"\nğŸ¯ ìš°ì„  ëª©í‘œ: 85% ë‹¬ì„± (í˜„ì¬: {current_accuracy:.1f}%)")
        print("   â€¢ ê³¼ì í•©ë³´ë‹¤ëŠ” ì ì ˆí•œ í•™ìŠµì— ì§‘ì¤‘")
        print("   â€¢ ëª¨ë¸ ìš©ëŸ‰ì„ ì•½ê°„ ì¦ê°€ ê³ ë ¤")
    elif current_accuracy < 92:
        print(f"\nğŸ¯ ìš°ì„  ëª©í‘œ: 92% ë‹¬ì„± (í˜„ì¬: {current_accuracy:.1f}%)")
        print("   â€¢ ê³¼ì í•© ë°©ì§€ì™€ ì„±ëŠ¥ í–¥ìƒì˜ ê· í˜•")
        print("   â€¢ ì ì§„ì  ì •ê·œí™” ê°•í™”")
    else:
        print(f"\nğŸ¯ ê³ ì„±ëŠ¥ ìœ ì§€: {current_accuracy:.1f}% ìœ ì§€í•˜ë©° ê³¼ì í•© í•´ê²°")
        print("   â€¢ ì„±ëŠ¥ ì†ì‹¤ ìµœì†Œí™”í•˜ë©° ì•ˆì •í™”")

# 9. ë©”ì¸ ì‹¤í–‰ í•¨ìˆ˜
def main():
    """ê³¼ì í•© í•´ê²° ë©”ì¸ í•¨ìˆ˜"""

    print("ğŸ›¡ï¸ ê³¼ì í•© í•´ê²°ì„ ìœ„í•œ ê°œì„ ëœ ëª¨ë¸ í•™ìŠµ")
    print("=" * 80)
    print("ğŸ¯ ëª©í‘œ: ê±´ì „í•œ Train/Val ì„±ëŠ¥ ê· í˜• ë‹¬ì„±")
    print("=" * 80)

    data_path = '/content/drive/MyDrive/withwe/011_AI ê¸°ë°˜ ì£¼ì–¼ë¦¬ ì´ë¯¸ì§€ ìë™ ë¶„ë¥˜ ëª¨ë¸ ê°œë°œ/accesary_datasets'

    if not os.path.exists(data_path):
        print(f"âŒ ë°ì´í„° ê²½ë¡œê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {data_path}")
        return

    try:
        # 1ë‹¨ê³„: ê³¼ì í•© ë°©ì§€ í•™ìŠµ
        print("\nğŸ¯ 1ë‹¨ê³„: ê³¼ì í•© ë°©ì§€ ê°•í™” í•™ìŠµ")
        model, train_losses, train_accs, val_losses, val_accs, best_acc, class_names = run_anti_overfitting_training(
            data_path=data_path,
            num_epochs=40  # ë” ì ì€ ì—í¬í¬ë¡œ ê³¼ì í•© ë°©ì§€
        )

        print(f"\nâœ… 1ë‹¨ê³„ ì™„ë£Œ! ìµœê³  ê±´ì „í•œ ì„±ëŠ¥: {best_acc:.2f}%")

        # 2ë‹¨ê³„: ê³¼ì í•© ì§„ë‹¨
        print("\nğŸ” 2ë‹¨ê³„: ê³¼ì í•© ì§„ë‹¨ ë¶„ì„")
        status, gap = diagnose_overfitting(train_losses, train_accs, val_losses, val_accs)

        # 3ë‹¨ê³„: í•´ê²°ì±… ì œì•ˆ
        print("\nğŸ’¡ 3ë‹¨ê³„: ê³¼ì í•© í•´ê²° ë°©ì•ˆ ì œì•ˆ")
        suggest_overfitting_solutions(gap, best_acc)

        # 4ë‹¨ê³„: ìµœì¢… í‰ê°€ (ê±´ì „í•œ ëª¨ë¸ì´ ìˆì„ ê²½ìš°)
        if gap <= 8:  # ê±´ì „í•œ ëª¨ë¸ì´ ì €ì¥ë˜ì—ˆë‹¤ë©´
            print("\nğŸ¯ 4ë‹¨ê³„: ìµœì¢… ê±´ì „í•œ ëª¨ë¸ í‰ê°€")

            # ëª¨ë¸ ë¡œë“œ
            device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
            model.load_state_dict(torch.load('anti_overfitting_best_model.pth'))

            # ê²€ì¦ ë°ì´í„°ë¡œ í‰ê°€
            val_transform = AntiOverfittingTransforms.get_val_transform(256)
            val_dataset = ImageFolder(root=os.path.join(data_path, 'val'), transform=val_transform)
            val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

            model.eval()
            all_preds = []
            all_labels = []

            print("ğŸ“Š ìµœì¢… ëª¨ë¸ í‰ê°€ ì¤‘...")
            with torch.no_grad():
                for images, labels in tqdm(val_loader):
                    images, labels = images.to(device), labels.to(device)
                    with autocast():
                        outputs = model(images)
                    _, predicted = torch.max(outputs, 1)
                    all_preds.extend(predicted.cpu().numpy())
                    all_labels.extend(labels.cpu().numpy())

            final_accuracy = 100 * np.mean(np.array(all_preds) == np.array(all_labels))

            print(f"\nğŸ¯ ìµœì¢… ê±´ì „í•œ ëª¨ë¸ ì„±ëŠ¥: {final_accuracy:.2f}%")

            # í˜¼ë™ í–‰ë ¬
            cm = confusion_matrix(all_labels, all_preds)
            plt.figure(figsize=(10, 8))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                        xticklabels=class_names, yticklabels=class_names)
            plt.title(f'ğŸ¯ Healthy Model Confusion Matrix\nAccuracy: {final_accuracy:.2f}%',
                     fontsize=16, fontweight='bold')
            plt.xlabel('Predicted Label')
            plt.ylabel('True Label')
            plt.show()

            # í´ë˜ìŠ¤ë³„ ì„±ëŠ¥
            report = classification_report(all_labels, all_preds, target_names=class_names,
                                         digits=3, output_dict=True)
            print(f"\nğŸ“ˆ í´ë˜ìŠ¤ë³„ ìµœì¢… ì„±ëŠ¥:")
            print("-" * 60)
            for class_name in class_names:
                metrics = report[class_name]
                print(f"{class_name:12} | P: {metrics['precision']:.3f} | "
                      f"R: {metrics['recall']:.3f} | F1: {metrics['f1-score']:.3f}")

        # ìµœì¢… ìš”ì•½
        print(f"\n" + "ğŸ‰" * 25)
        print(f"ğŸ›¡ï¸ ê³¼ì í•© í•´ê²° ê²°ê³¼ ìš”ì•½")
        print(f"ğŸ‰" * 25)
        print(f"ğŸ“Š ì§„ë‹¨ ê²°ê³¼: {status}")
        print(f"ğŸ“ˆ ì„±ëŠ¥ ì°¨ì´: {gap:+.2f}% (ëª©í‘œ: â‰¤8%)")
        print(f"ğŸ† ë‹¬ì„± ì„±ëŠ¥: {best_acc:.2f}%")
        print(f"ğŸ’¾ ì €ì¥ëœ ëª¨ë¸: anti_overfitting_best_model.pth")

        if gap <= 5:
            print(f"âœ… ì„±ê³µ: ê±´ì „í•œ ëª¨ë¸ ë‹¬ì„±!")
        elif gap <= 8:
            print(f"âš ï¸ ì–‘í˜¸: ê²½ë¯¸í•œ ê³¼ì í•©, ì¶”ê°€ ê°œì„  ê¶Œì¥")
        else:
            print(f"ğŸš¨ ì£¼ì˜: ì‹¬ê°í•œ ê³¼ì í•©, ì„¤ì • ì¬ì¡°ì • í•„ìš”")

    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {str(e)}")
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        gc.collect()

# 10. ì¶”ê°€ ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜ë“¤
def quick_overfitting_check(train_acc, val_acc):
    """ë¹ ë¥¸ ê³¼ì í•© ì²´í¬"""
    gap = val_acc - train_acc

    if gap > 10:
        return "ğŸš¨ ì‹¬ê°í•œ ê³¼ì í•©"
    elif gap > 5:
        return "âš ï¸ ê³¼ì í•© ì˜ì‹¬"
    elif gap > -2:
        return "âœ… ê±´ì „"
    else:
        return "ğŸ“‰ ì–¸ë”í”¼íŒ…"

def calculate_healthy_score(train_acc, val_acc):
    """ê±´ì „ì„± ì ìˆ˜ ê³„ì‚° (0-100)"""
    gap = abs(val_acc - train_acc)

    if gap <= 2:
        return 100
    elif gap <= 5:
        return 90 - (gap - 2) * 10
    elif gap <= 10:
        return 60 - (gap - 5) * 6
    else:
        return max(0, 30 - (gap - 10) * 3)

# ì‹¤í–‰
if __name__ == "__main__":
    main()
```
